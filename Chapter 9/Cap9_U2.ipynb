### Clula de memoria larga a corto plazo

#### Clula de memoria larga a corto plazo (LSTM)

* Mejora las clulas `RNN` normales y puede mantener una `correlacin duradera`.


* As, se ide un nuevo tipo de celda de memoria llamada `LSTM`.


* La celda LSTM tiene una estructura interna bastante `complicada`.

* $c_t$ = estado de celda y   = funcin sigmoidea

* Internamente, existen `compuertas` que operan con las funciones `sigmoidea` y `tangente hiperblica` (tanh).
