### Célula de memoria larga a corto plazo

#### Célula de memoria larga a corto plazo (LSTM)

* Mejora las células `RNN` normales y puede mantener una `correlación duradera`.


* Así, se ideó un nuevo tipo de celda de memoria llamada `LSTM`.


* La celda LSTM tiene una estructura interna bastante `complicada`.

SIC2023_C9_U2_P078.png

* $c_t$ = “estado de celda” y σ  = función sigmoidea

* Internamente, existen `compuertas` que operan con las funciones `sigmoidea` y `tangente hiperbólica` (tanh).

* Las células LSTM pueden formar una `secuencia` como las células RNN normales.
