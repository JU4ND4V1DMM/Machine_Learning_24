## Pipelines (Canalización)

### Perspectiva estructural de los tipos de datos

* Los datos recopilados, desde una perspectiva estructural (estructura de esquema o computabilidad), se pueden dividir en tres categorías: datos estructurados, no estructurados y semiestructurados.
 * Los datos estructurados se refieren a datos que tienen una forma basada en la estructura del esquema estructurado (formulario) y se almacenan en campos fijos como RDB (Relational Data Base) y hoja de cálculo y son consistentes en valor y formato.
 * Los datos no estructurados se refieren a datos que no tienen una estructura de esquema y no se almacenan en campos fijos como redes sociales, tableros de anuncios web y NoSQL.
 * Los datos semiestructurados se refieren a datos que tienen una estructura de esquema (formal) y contienen metadatos como XML, HTML, blogs, registros del sistema y alarmas, y son inconsistentes en valor y formato.

### Series

#### Una Serie es una forma de matrices unidimensionales enumeradas secuencialmente.

* Como es una matriz, todos los elementos de la serie deben pertenecer a un mismo tipo de datos.
* Se pueden crear series simples a partir de secuencias como listas, tuplas y vectores.
* La palabra "serie" es singular Y plural. Su origen latino, 'serere', significa unir o conectar.
* El valor predeterminado de la serie es un índice entero. La etiqueta del primer elemento es 0, la etiqueta del segundo elemento es 1 y aumenta de esta manera.
* El valor del atributo de la serie es la lista de todos los valores de la serie.
* El valor del atributo del índice significa un índice de una serie.
* El valor del atributo de `index.values` es un vector de todos los valores indexados.

* En el ejemplo anterior, el orden de los nombres de los sujetos proporcionados por índice es diferente. Sin embargo, pandas busca y clasifica el mismo nombre de sujeto name(index) y agrega puntuaciones del mismo nombre de sujeto name(index).

* Si el número de elementos en las dos series es diferente o el tamaño de la serie es el mismo, se trata como si no fuera un número Not a Number (NaN), lo que significa que no hay un valor válido.

```python
estudiante1 = pd.Series({'kor':np.nan,'eng':80,'math':90})
estudiante2 = pd.Series({'math':80,'kor':90})

adicion = estudiante1 + estudiante2
print(f'adicion: {adicion}')
resta = estudiante1 - estudiante2
print(f'resta: {resta}')
multiplicacion = estudiante1 * estudiante2
print(f'multiplicacion: {multiplicacion}')
division = estudiante1 / estudiante2
print(f'division: {division}')

result = pd.DataFrame([adicion, resta, multiplicacion, division],
                      index=['adicion', 'resta', 'multiplicacion', 'division'])
result
```

#### Métodos de operación

* Como has aprendido en diapositivas anteriores, cuando no hay un índice común o NAN en los valores, devuelve NaN. Para que no se produzca este resultado, debemos usar el valor de rellenar (fill) para el conjunto de datos.

```python
estudiante1 = pd.Series({'kor':np.nan,'eng':80,'math':90})
estudiante2 = pd.Series({'math':80,'kor':90})

sr_add = estudiante1.add(estudiante2, fill_value=0)
sr_sub = estudiante1.sub(estudiante2, fill_value=0)
sr_mul = estudiante1.mul(estudiante2, fill_value=0)
sr_div = estudiante1.div(estudiante2, fill_value=0)

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index=['adicion', 'resta', 'multiplicacion', 'division'])
result
```

### Agregado
Los objetos de marco de datos se generan de varias maneras, pero la forma óptima es usar diccionarios en listas de la misma longitud o usar matrices Numpy. El índice del marco de datos creado se sustituye automáticamente de la misma manera que la serie, y las columnas se ordenan y almacenan.


```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
frame
```

Con el método del encabezado, solo se pueden designar o imprimir cinco filas.

```python
frame.head()
```

**SI** las columnas se designan en el orden deseado, los objetos del DataFrame se mejoran en consecuencia.


```python
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

Al igual que la serie, si pasa un valor que no está en el diccionario, se almacena como un valor faltante.


```python
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'],
                      index=['one', 'two', 'three', 'four',
                             'five', 'six'])
```

```python
frame2
```

```python
frame2.columns
```

Las columnas de DataFrame, como las Series, se pueden abordar en notación de diccionario o formato de atributos.


```python
frame2['state']
```

```python
frame2.year
```

Si el objeto de serie devuelto tiene un índice como un DataFrame, el atributo de nombre se completa con el valor apropiado. Se puede acceder a las filas por nombre mediante un atributo loc o ubicación.

```python
frame2.loc['three']
```

Las columnas se pueden sustituir. Por ejemplo, se puede sustituir un valor escalar o un valor de matriz en la columna "debt" (deuda) actualmente vacía.


```python
frame2['debt'] = 16.5
frame2
```

```python
frame2['debt'] = np.arange(6.)
frame2
```

Al sustituir una lista o matriz en una columna, la longitud del valor a sustituir debe ser igual al tamaño del DataFrame. Cuando se sustituye una serie, los valores se sustituyen según el índice del DataFrame y los valores faltantes se sustituyen por índices que no existen.

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
frame2['debt'] = val
frame2
```

```python
# Crea series de pandas con datos del diccionario.
student1 = pd.Series({'eng':100, 'math':80, 'science':90})
student2 = pd.Series({'science':80, 'eng':90, 'math':80})
print(student1)
print('\n')
print(student2)
print('\n')
```

```python
# Realiza operaciones aritméticas con las puntuaciones de cada materia de los dos estudiantes.
addition = student1 + student2               #adición
subtraction = student1 - student2            #resta
multiplication = student1 * student2         #multiplicación
division = student1 / student2               #división
print(type(division))
```

```python
# Combinar los resultados de la operación aritmética en un DataFrame (Series -> DataFrame)
result = pd.DataFrame([addition, subtraction, multiplication, division],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
print(result)
```

```python
# Crea Series de pandas con datos del diccionario.
student1 = pd.Series({'eng':np.nan, 'math':80, 'science':90})
student2 = pd.Series({'science':80, 'eng':90})
print(student1)
print('\n')
print(student2)
print('\n')
```

```python
# Realizar un cálculo aritmético con las puntuaciones de cada materia de los dos estudiantes (usando el método de operación)
sr_add = student1.add(student2, fill_value=0)   #adición
sr_sub = student1.sub(student2, fill_value=0)   #resta
sr_mul = student1.mul(student2, fill_value=0)   #multiplicación
sr_div = student1.div(student2, fill_value=0)   #división
```

```python
# Combinar los resultados de la operación aritmética en un DataFrame (Series -> DataFrame
result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
print(result)
```
