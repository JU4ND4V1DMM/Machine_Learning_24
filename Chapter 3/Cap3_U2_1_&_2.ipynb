## Pipelines (Canalizacin)

### Perspectiva estructural de los tipos de datos

* Los datos recopilados, desde una perspectiva estructural (estructura de esquema o computabilidad), se pueden dividir en tres categoras: datos estructurados, no estructurados y semiestructurados.
 * Los datos estructurados se refieren a datos que tienen una forma basada en la estructura del esquema estructurado (formulario) y se almacenan en campos fijos como RDB (Relational Data Base) y hoja de clculo y son consistentes en valor y formato.
 * Los datos no estructurados se refieren a datos que no tienen una estructura de esquema y no se almacenan en campos fijos como redes sociales, tableros de anuncios web y NoSQL.
 * Los datos semiestructurados se refieren a datos que tienen una estructura de esquema (formal) y contienen metadatos como XML, HTML, blogs, registros del sistema y alarmas, y son inconsistentes en valor y formato.

### Series

#### Una Serie es una forma de matrices unidimensionales enumeradas secuencialmente.

* Como es una matriz, todos los elementos de la serie deben pertenecer a un mismo tipo de datos.
* Se pueden crear series simples a partir de secuencias como listas, tuplas y vectores.
* La palabra "serie" es singular Y plural. Su origen latino, 'serere', significa unir o conectar.
* El valor predeterminado de la serie es un ndice entero. La etiqueta del primer elemento es 0, la etiqueta del segundo elemento es 1 y aumenta de esta manera.
* El valor del atributo de la serie es la lista de todos los valores de la serie.
* El valor del atributo del ndice significa un ndice de una serie.
* El valor del atributo de `index.values` es un vector de todos los valores indexados.

* Si el nmero de elementos en las dos series es diferente o el tamao de la serie es el mismo, se trata como si no fuera un nmero Not a Number (NaN), lo que significa que no hay un valor vlido.

```python
estudiante1 = pd.Series({'kor':np.nan,'eng':80,'math':90})
estudiante2 = pd.Series({'math':80,'kor':90})

adicion = estudiante1 + estudiante2
print(f'adicion: {adicion}')
resta = estudiante1 - estudiante2
print(f'resta: {resta}')
multiplicacion = estudiante1 * estudiante2
print(f'multiplicacion: {multiplicacion}')
division = estudiante1 / estudiante2
print(f'division: {division}')

result = pd.DataFrame([adicion, resta, multiplicacion, division],
                      index=['adicion', 'resta', 'multiplicacion', 'division'])
result
```

#### Mtodos de operacin

* Como has aprendido en diapositivas anteriores, cuando no hay un ndice comn o NAN en los valores, devuelve NaN. Para que no se produzca este resultado, debemos usar el valor de rellenar (fill) para el conjunto de datos.

```python
estudiante1 = pd.Series({'kor':np.nan,'eng':80,'math':90})
estudiante2 = pd.Series({'math':80,'kor':90})

sr_add = estudiante1.add(estudiante2, fill_value=0)
sr_sub = estudiante1.sub(estudiante2, fill_value=0)
sr_mul = estudiante1.mul(estudiante2, fill_value=0)
sr_div = estudiante1.div(estudiante2, fill_value=0)

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index=['adicion', 'resta', 'multiplicacion', 'division'])
result
```

### Agregado
Los objetos de marco de datos se generan de varias maneras, pero la forma ptima es usar diccionarios en listas de la misma longitud o usar matrices Numpy. El ndice del marco de datos creado se sustituye automticamente de la misma manera que la serie, y las columnas se ordenan y almacenan.


```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
frame
```

Con el mtodo del encabezado, solo se pueden designar o imprimir cinco filas.

```python
frame.head()
```

**SI** las columnas se designan en el orden deseado, los objetos del DataFrame se mejoran en consecuencia.


```python
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

Al igual que la serie, si pasa un valor que no est en el diccionario, se almacena como un valor faltante.


```python
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'],
                      index=['one', 'two', 'three', 'four',
                             'five', 'six'])
```

```python
frame2
```

```python
frame2.columns
```

Las columnas se pueden sustituir. Por ejemplo, se puede sustituir un valor escalar o un valor de matriz en la columna "debt" (deuda) actualmente vaca.


```python
frame2['debt'] = 16.5
frame2
```

```python
frame2['debt'] = np.arange(6.)
frame2
```

Al sustituir una lista o matriz en una columna, la longitud del valor a sustituir debe ser igual al tamao del DataFrame. Cuando se sustituye una serie, los valores se sustituyen segn el ndice del DataFrame y los valores faltantes se sustituyen por ndices que no existen.

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
frame2['debt'] = val
frame2
```

```python
# Crea series de pandas con datos del diccionario.
student1 = pd.Series({'eng':100, 'math':80, 'science':90})
student2 = pd.Series({'science':80, 'eng':90, 'math':80})
print(student1)
print('\n')
print(student2)
print('\n')
```

```python
# Realiza operaciones aritmticas con las puntuaciones de cada materia de los dos estudiantes.
addition = student1 + student2               #adicin
subtraction = student1 - student2            #resta
multiplication = student1 * student2         #multiplicacin
division = student1 / student2               #divisin
print(type(division))
```

```python
# Combinar los resultados de la operacin aritmtica en un DataFrame (Series -> DataFrame)
result = pd.DataFrame([addition, subtraction, multiplication, division],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
print(result)
```

```python
# Crea Series de pandas con datos del diccionario.
student1 = pd.Series({'eng':np.nan, 'math':80, 'science':90})
student2 = pd.Series({'science':80, 'eng':90})
print(student1)
print('\n')
print(student2)
print('\n')
```

```python
# Realizar un clculo aritmtico con las puntuaciones de cada materia de los dos estudiantes (usando el mtodo de operacin)
sr_add = student1.add(student2, fill_value=0)   #adicin
sr_sub = student1.sub(student2, fill_value=0)   #resta
sr_mul = student1.mul(student2, fill_value=0)   #multiplicacin
sr_div = student1.div(student2, fill_value=0)   #divisin
```

```python
# Combinar los resultados de la operacin aritmtica en un DataFrame (Series -> DataFrame
result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
print(result)
```
