#### **Qu es NumPy?**
**NumPy significa Numerical Python, y es un paquete bsico para ciencia de datos.**

* Admite transformaciones discretas de Fourier, lgebra lineal bsica, operaciones estadsticas bsicas, simulaciones aleatorias, etc.

* El objeto **ndarray** es el ncleo del paquete NumPy. Procesa matrices de n dimensiones de tipos de datos homogneos.


#### Distinguir las diferencias entre vectores NumPy y secuencias estndar de Python <br> (Listas, Tuplas, Diccionarios, Conjuntos).
* Los vectores NumPy tienen un tamao fijo cuando se generan a diferencia de las listas de Python (que pueden crecer dinmicamente).
* Un cambio en el tamao de **ndarray** crea un nuevo vector y elimina el original.

* Sin embargo, las listas permiten vectores de varios tamaos.
* Los vectores NumPy calculan fcilmente operaciones matemticas avanzadas y de otro tipo en grandes cantidades de datos.
* Normalmente, estas operaciones se realizan de forma ms eficiente y con menos cdigo que cuando se realizan con las secuencias integradas de Python.
