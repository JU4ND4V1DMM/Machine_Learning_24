#### Tipo de datos

* En primer lugar, el tipo de datos, en informática y lenguajes de programación, es una clasificación que identifica un tipo de datos como flotante, entero, booleano, caracter y cadena, al mismo tiempo que determina el tamaño de los datos. Debe ser utilizado por el tipo de dato aplicable. Los tipos de datos difieren según el lenguaje de programación, pero la mayoría de los conceptos de tipos de datos del lenguaje C se heredan y utilizan y han afectado a otros lenguajes.
* Almacenar datos en la memoria tiene un alto costo operacional, por lo que lenguajes como C fueron diseñados para optimizar y almacenar datos usando el menor espacio posible. Este diseño introdujo un nuevo problema en el manejo del tamaño y manejo de los datos.
* Por ejemplo, en C, los tipos de datos enteros se dividen en char (compatible con números enteros), short, int y long. Cada tamaño de byte difiere de los otros tipos de datos.

#### Estructura de datos
* En segundo lugar, la estructura de datos en informática se refiere a la organización, gestión y almacenamiento de los datos que permite un acceso y modificación eficientes. Depende, en gran medida, de la estructura de datos para encontrarlos más rápidamente. En otras palabras, encontrar los datos es una cuestión de algoritmos (que se explicará más adelante), por lo que se puede decir que una buena estructura de datos es una condición necesaria y suficiente para un buen algoritmo.
* Para ser más específico, la estructura de datos se refiere a un grupo de valores de datos, una relación entre datos y una función o comando aplicable a los datos. Las estructuras de datos cuidadosamente seleccionadas permiten utilizar algoritmos más eficientes.
* Una estructura de datos diseñada eficazmente permite realizar operaciones con recursos mínimos, como son el tiempo de ejecución o la capacidad de memoria.
* Existen varios tipos de estructuras de datos, cada una de las cuales se adapta a cada operación y propósito.
* Al diseñar varios programas, la prioridad debe ser considerar y seleccionar la estructura de datos más adecuada. Esto se debe a que cuando se fabrica un sistema a gran escala, la dificultad de implementación y el rendimiento del producto final dependen en gran medida de la estructura de datos.
* Una vez seleccionada la estructura de datos, queda relativamente claro cuál es el algoritmo que se debe aplicar. Hay ocasiones en las que este orden se invierte: cuando la operación de destino necesariamente requiere un algoritmo particular, y el algoritmo dado produce el mejor rendimiento con la estructura de datos particular. En cualquier caso, es fundamental seleccionar una estructura de datos adecuada.

#### **¿Qué es NumPy?**
**NumPy significa Numerical Python, y es un paquete básico para ciencia de datos.**

* Es una librería de Python que proporciona objetos multidimensionales de vector, varios objetos derivados (matrices, etc.) y una variedad de rutinas para una operación rápida en vectores y matrices.

* Admite transformaciones discretas de Fourier, álgebra lineal básica, operaciones estadísticas básicas, simulaciones aleatorias, etc.

* El objeto **ndarray** es el núcleo del paquete NumPy. Procesa matrices de n dimensiones de tipos de datos homogéneos.


#### Distinguir las diferencias entre vectores NumPy y secuencias estándar de Python <br> (Listas, Tuplas, Diccionarios, Conjuntos).
* Los vectores NumPy tienen un tamaño fijo cuando se generan a diferencia de las listas de Python (que pueden crecer dinámicamente).
* Un cambio en el tamaño de **ndarray** crea un nuevo vector y elimina el original.
* Los elementos del vector NumPy son todos tipos de datos homogéneos; por tanto, mantienen el mismo tamaño en la memoria.
* Sin embargo, las listas permiten vectores de varios tamaños.
* Los vectores NumPy calculan fácilmente operaciones matemáticas avanzadas y de otro tipo en grandes cantidades de datos.
* Normalmente, estas operaciones se realizan de forma más eficiente y con menos código que cuando se realizan con las secuencias integradas de Python.
* Cada vez más paquetes científicos y matemáticos basados en Python utilizan vectores NumPy. Por lo general, estos admiten la entrada secuencial de Python, pero convierten dicha entrada en vectores NumPy antes de procesarlas y, a menudo, generan como salida vectores NumPy.
* En otras palabras, no es suficiente confiar únicamente en las secuencias integradas de Python para utilizar eficientemente el software científico/matemático actual basado en Python. También es necesario tener conocimientos de vectores NumPy para aumentar la eficiencia.


#### ¿Porqué usar Vectores NumPy? (1/3)

* Anteriormente, aprendimos que Python no admite el uso de vectores y matrices de forma dirécta, pero permite la organización de estructuras de datos a través de la librería NumPy.
* Los conceptos de organización aprendidos en los lenguajes de programación estándar no son muy diferentes en Python.
* En programación, el método para aumentar el rendimiento computacional es disminuyendo el uso de bucles.
* Los bucles en cálculos a gran escala requieren un cálculo por repetición, lo que provoca un rendimiento deficiente.
* Los vectores NumPy permiten una amplia variedad de operaciones de procesamiento de datos a través de operaciones matriciales concisas en lugar de bucles.
* Este uso de operaciones matriciales computacionales para eliminar explícitamente los bucles se denomina **vectorización**. Las operaciones matemáticas para matrices vectorizadas suelen ser dos o tres, si no diez o incluso cien veces más rápidas que las operaciones puras de Python. La **transmisión**, otro método que aprenderemos más adelante, es una operación vectorial muy poderosa.
