$\sigma^2 = \sum{\frac{(x - \mu)^2}{N}}$

donde:

 es la varianza <br>
x es un valor de la variable aleatoria: (x) <br>
 es la media de la variable aleatoria: (x_mean) <br>
N es el nmero de elementos en la poblacin: (x.size)

#### lgebra lineal en vectores NumPy

* NumPy admite no solo operaciones simples de matrices sino tambin operaciones matriciales (matriz bidimensional) para lgebra lineal.
* Entre los diversos mtodos, veamos primero cmo obtener la multiplicacin de matrices, la matriz transpuesta, la matriz inversa y el determinante.

#### Operaciones matriciales

* El siguiente es el mtodo para obtener el producto matricial, la matriz transpuesta, la matriz inversa y el determinante de las matrices A y B.
- producto matricial
- transponer matriz
- matriz inversa
- determinante

* Para operaciones matriciales, crearemos las matrices A y B de 2x2 de la siguiente manera:

```python
A = np.array([0,1,2,3]).reshape(2,2)
A
```

```python
B = np.array([3,2,0,1]).reshape(2,2)
B
```

Ejemplo del producto de las matrices A y B (A*B). Se pueden utilizar ambos mtodos.

```python
A.dot(B)
```

```python
np.dot(A,B)
```

El siguiente es un ejemplo de cmo encontrar la matriz transpuesta de la matriz A. Se pueden utilizar ambos mtodos.

```python
np.transpose(A)
```

```python
A.transpose()
```

Cmo encontrar la matriz inversa de la matriz A?

```python
np.linalg.inv(A)
```

Un ejemplo de obtencin de un determinante de la matriz A.

```python
np.linalg.det(A)
```
