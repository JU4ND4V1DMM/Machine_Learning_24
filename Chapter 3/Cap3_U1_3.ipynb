$ </>$ **Línea [38]:** <br>
La suma de las desviaciones al cuadrado dividida por el número de elementos en el vector se llama varianza.

La fórmula para calcular la varianza de una población es:

$\sigma^2 = \sum{\frac{(x - \mu)^2}{N}}$

donde:

σ² es la varianza <br>
x es un valor de la variable aleatoria: (x) <br>
μ es la media de la variable aleatoria: (x_mean) <br>
N es el número de elementos en la población: (x.size)

Este resultado se puede calcular directamente mediante el proceso anterior, o se puede utilizar var().

```python
x.var()
```

```python
np.sqrt(x.var())
```

#### Álgebra lineal en vectores NumPy

* NumPy admite no solo operaciones simples de matrices sino también operaciones matriciales (matriz bidimensional) para álgebra lineal.
* Entre los diversos métodos, veamos primero cómo obtener la multiplicación de matrices, la matriz transpuesta, la matriz inversa y el determinante.

#### Operaciones matriciales

* El siguiente es el método para obtener el producto matricial, la matriz transpuesta, la matriz inversa y el determinante de las matrices A y B.
- producto matricial
- transponer matriz
- matriz inversa
- determinante

* Para operaciones matriciales, crearemos las matrices A y B de 2x2 de la siguiente manera:

```python
A = np.array([0,1,2,3]).reshape(2,2)
A
```

```python
B = np.array([3,2,0,1]).reshape(2,2)
B
```

Ejemplo del producto de las matrices A y B (A*B). Se pueden utilizar ambos métodos.

```python
A.dot(B)
```

```python
np.dot(A,B)
```

El siguiente es un ejemplo de cómo encontrar la matriz transpuesta de la matriz A. Se pueden utilizar ambos métodos.

```python
np.transpose(A)
```

```python
A.transpose()
```

¿Cómo encontrar la matriz inversa de la matriz A?

```python
np.linalg.inv(A)
```

Un ejemplo de obtención de un determinante de la matriz A.

```python
np.linalg.det(A)
```
