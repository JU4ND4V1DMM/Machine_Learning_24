#### Clasificación de los datos

* Los datos continuos y los datos categóricos son los tipos más básicos de datos estructurados.
* Los datos continuos incluyen datos de continuidad, como la velocidad y la duración, y datos discretos, como la frecuencia de ocurrencia de eventos.
* Los datos categóricos se refieren al nombre de las ciudades de cada país (Washington, Nueva York, Los Ángeles) o al tipo de coche (bus, taxi, camión)
* Los datos binarios son un caso especial entre los tipos categóricos, de tener cualquier valor, como pueden ser 0 y 1, sí/no o verdadero/falso,
* Entre los tipos categóricos, las calificaciones (1,2,3,4,5) en las que se clasifican los valores dentro de la categoría se denominan datos ordinales.


#### ¿Por qué clasificar los datos?

* El software de ciencia de datos como Python utiliza esta información de tipo de datos para el rendimiento computacional.
* El tipo de datos determina cómo realizar el cálculo relacionado con la variable.
* Por ejemplo, en R o Python, los datos ordinales se clasifican en factores ordenados y se utilizan para mantener el orden deseado por los usuarios en gráficos, tablas y modelos estadísticos.

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig0=plt.figure()
axes0 = fig0.add_axes([0,0,1,1])             # Izquierda, abajo, ancho, alto.
axes0.plot(x,y,color='red',linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

#### 1.3. Múltiples gráficos en ejes separados:

```python
fig0=plt.figure(figsize=(8,3), dpi=80)       # Configuración de ancho, alto y DPI.
axes1 = fig0.add_axes([0,0,0.3,1])            # Izquierda, abajo, ancho, alto.
axes2 = fig0.add_axes([0.5,0,0.3,1])          # Izquierda, abajo, ancho, alto.
axes1.plot(x,y,color='red',linestyle=':')
axes2.plot(x,z,color='blue',linestyle='-.')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('SINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('COSINE')
plt.show()
```

#### 1.4. Múltiples gráficos en una serie de ejes:

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
# (0,0)
c = np.random.randn(10000)
axes[0,0].hist(c,bins=100,color='green',density=True,alpha=0.3)
axes[0,0].set_title('HISTOGRAM')
# (1,0)
axes[1,0].plot(x,z,linestyle='none',marker='^',markersize=10,markerfacecolor='purple',markeredgecolor='red',alpha=0.5)
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,0].set_title('COSINE')
# (0,1)
a = np.array(['Q1', 'Q2', 'Q3','Q4'])
b = np.array([ 234.0, 254.7, 144.6, 317.6])
axes[0,1].bar(a,b,color='blue',alpha=0.5)
axes[0,1].set_title('BAR PLOT')
# (1,1)
axes[1,1].plot(x,y,color='green',linestyle=':')
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].set_title('SINE')

plt.tight_layout()                             # Evite la superposición.
plt.show()
```
