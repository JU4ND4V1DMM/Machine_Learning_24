### Cortes Vectoriales

#### En lugar de seleccionar un elemento mediante la indexación, el corte selecciona una parte del vector o la matriz especificando un rango.

* Para una matriz unidimensional, el corte especifica las posiciones del principio y el final, como se muestra a continuación.
* Matriz[posición_inicial]:posición_final]
* Si no se especifica la posición inicial, la posición inicial pasa a ser 0 y el rango pasa a ser **“0 hasta la posición final -1”**. Si no se especifica la posición final, la "posición final" se convierte en la longitud de la matriz y el rango se convierte en **"posición_inicial hasta el final de la matriz"**.

```python
import numpy as np
```

```python
b1=np.array([0,10,20,30,40,50])
b1[1:4]
```

#### Cortes
* En una matriz unidimensional, el corte se realiza sin especificar una "posición_inicial" y una "posición_final", como se muestra a continuación.

```python
b1[:3]
```

```python
b1[2:]
```

* Ahora, consideremos el caso de una matriz bidimensional.

```python
arr2d=np.arange(1,10).reshape(3,3)
arr2d
```

* Intentémos cortar la matriz de arriba.

```python
arr2d[:2]
```

$ </>$ **Línea [25]:** <br>
Elijamos desde el principio hasta la segunda línea de "arr2d".

* También es posible dividir vectores multidimensionales cruzando varios índices.

```python
arr2d[:2,1:]
```

* Como se muestra arriba, el corte siempre brinda una vista de la matriz en la misma dimensión. Se pueden utilizar juntos un índice entero y un segmento para obtener un segmento de menor dimensión.
* Por ejemplo, si deseas seleccionar solo las dos primeras columnas de la segunda fila, debes hacer lo siguiente.

```python
arr2d[1,:2]
```

* Si quieres seleccionar solo la tercera columna en las dos primeras filas, debes hacer lo siguiente.

```python
arr2d[:2,2]
```

* Si solo se usan dos puntos, elegimos el eje completo, por lo que retorna la porción de la dimensión original.

```python
arr2d[:,:1]
```

```python
arr2d
```

```python
arr2d[2]
```

```python
arr2d[2, :]
```

```python
arr2d[2:,:]
```

### Indexación elegante

#### Devuelve una nueva matriz de forma y tipo determinados sin inicializar entradas. Puedes consultar ayuda help(np.empty) para obtener más detalles.

```python
np.empty((8,4))
```

```python
arr=np.empty((8,4))
```

```python
for i in range(8):
  arr[i]=i
```

```python
arr
```

* Seleccionando filas en un orden específico podemos omitir el ndarrray o la lista que contiene el orden deseado.

```python
arr[[4,3,0,6]]
```

* Si utilizamos números negativos como índice, la selección de fila se realiza contando desde el final.<br>arr[[-3,-5]] indica que cuente primero desde la última fila hacia el comienzo 3 veces. Es decir, primero fila de 7 (va 1), luego fila de 6 (va 2) y seleccione fila de 5 (que es la 3ra). <br>
Luego, el segundo índice sigue la misma secuencia y por tanto al contar 5 del final hacia el principio la fila que selecciona es la fila de 3.

```python
arr[[-3,-5]]
```

* Para indexar valores que son sólo múltiplos de 5: podemos especificar una condición booleana dentro de la variable para obtenerla.

```python
arr2=np.arange(20)
arr2
```

```python
arrMask=(arr2%5==0)
arrMask
```

```python
arr2[arrMask]
```

* También se puede obtener combinando dos condiciones.

```python
arr2>10
```

```python
arrMask2=(arr2>=10)
```

```python
arr2[arrMask & arrMask2]
```

```python
for i in arr2:    #recorra todo el vector arr2
  # si arrMask[indice] es verdadero y arrMask2[indice] también
  if arrMask[i] and arrMask2[i]: # Ref a Linea 49
    print(i,",",arrMask[i])  # imprima el índice y resultado de arrMask[indice]
  else:  # de lo contrario
    print(i,",")  # imprima solo el índice
```
