### Cortes Vectoriales

#### En lugar de seleccionar un elemento mediante la indexacin, el corte selecciona una parte del vector o la matriz especificando un rango.

* Para una matriz unidimensional, el corte especifica las posiciones del principio y el final, como se muestra a continuacin.
* Matriz[posicin_inicial]:posicin_final]
* Si no se especifica la posicin inicial, la posicin inicial pasa a ser 0 y el rango pasa a ser **0 hasta la posicin final -1**. Si no se especifica la posicin final, la "posicin final" se convierte en la longitud de la matriz y el rango se convierte en **"posicin_inicial hasta el final de la matriz"**.

```python
import numpy as np
```

```python
b1=np.array([0,10,20,30,40,50])
b1[1:4]
```

#### Cortes
* En una matriz unidimensional, el corte se realiza sin especificar una "posicin_inicial" y una "posicin_final", como se muestra a continuacin.

```python
b1[:3]
```

```python
b1[2:]
```

* Ahora, consideremos el caso de una matriz bidimensional.

```python
arr2d=np.arange(1,10).reshape(3,3)
arr2d
```

* Intentmos cortar la matriz de arriba.

```python
arr2d[:2]
```

$ </>$ **Lnea [25]:** <br>
Elijamos desde el principio hasta la segunda lnea de "arr2d".

* Tambin es posible dividir vectores multidimensionales cruzando varios ndices.

```python
arr2d[:2,1:]
```

* Como se muestra arriba, el corte siempre brinda una vista de la matriz en la misma dimensin. Se pueden utilizar juntos un ndice entero y un segmento para obtener un segmento de menor dimensin.
* Por ejemplo, si deseas seleccionar solo las dos primeras columnas de la segunda fila, debes hacer lo siguiente.

```python
arr2d[1,:2]
```

* Si quieres seleccionar solo la tercera columna en las dos primeras filas, debes hacer lo siguiente.

```python
arr2d[:2,2]
```

* Si solo se usan dos puntos, elegimos el eje completo, por lo que retorna la porcin de la dimensin original.

```python
arr2d[:,:1]
```

```python
arr2d
```

```python
arr2d[2]
```

```python
arr2d[2, :]
```

```python
arr2d[2:,:]
```
