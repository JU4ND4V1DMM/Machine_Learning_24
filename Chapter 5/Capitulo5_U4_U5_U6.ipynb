$\color{blue}{\text{|}}$ **Proceso de divisin repetitivo**


**1)** Selecciona $x_i$, una de las variables, y el valor $x_i$ ($s_i$ como criterio de divisin) se designa para dividir el espacio de dimensin $p$ en dos.

**2)** $x_i \to \{x_i <= s_i\} \cup \{x_i > s_i\}$

**3)** Selecciona una variable de nuevo y divdela de la misma manera.

**4)** Repite el proceso hasta que alcance la pureza deseada.

Entropa: Grado de incertidumbre

La entropa es una funcin de estado utilizada para explicar el flujo de energa termodinmicamente no til (energa que no se puede convertir en trabajo).

Estadsticamente, la entropa es un logaritmo del nmero de estado microscpico que responde al estado macroscpico. En general, la entropa no se conserva. Aument a travs del tiempo de acuerdo con la segunda ley de la termodinmica.

#### $\color{blue}{\text{|}}$ **Medida de Impureza**

$\color{blue}{\to}$ Selecciona los nodos secundarios con una variable predictora que reduce el ndice de Gini y la creacin de particiones ptima.

$\color{blue}{\to}$ Si el conjunto de datos $T$ se divide en $k$ categoras y los coeficientes de rendimiento de la categora son $p_1,\cdots, p_k$, se expresa como la siguiente ecuacin.


\begin{align}
\large
Gini(T)=1-\sum_{l=1}^kp_l^2
\end{align}

SIC2023_C5_U4_324.png

#### $\color{blue}{\text{|}}$ **Medicin de Entropa**

$\color{blue}{\to}$ En termodinmica, la entropa mide el grado de desorden.

$\color{blue}{\to}$ Crea nodos secundarios con una variable predictora con la medida de entropa ms baja y la creacin de particiones ptima.

$\color{blue}{\to}$ Si el conjunto de datos $T$ se divide en $k$ categoras y los coeficientes de rendimiento de la categora son $p_1, \cdots, p_k$, se expresa como la siguiente ecuacin.


\begin{align}
\large
Entropa(T)=-\sum_{l=1}^kp_llog_2p_l
\end{align}


<br>

**Ejemplo.** Si 4 categoras consisten en proporciones de 0,25, 0,25, 0,25, 0,25 ($T_0$):

\begin{align}
\large
Entropa(T_0)=-(0.25log_20.25) \times 4 = 2
\end{align}

<br>

**Ejemplo.** Si 4 categoras consisten en proporciones de 0.5, 0.25, 0.25, 0 ($T_1$):

\begin{align}
\large
Entropa(T_1)=-(0.5log_20.5+0.25log_20.25+0.25log_20.25)=1.5
\end{align}

#### $\color{blue}{\text{|}}$ **Arbol de Regresin**

$\color{blue}{\to}$ El espacio configuracional se divide en regiones: $\{R_1,R_2, ...,R_J\}$.

$\color{blue}{\to}$ La estructura de rbol se entrena minimizando el RSS (suma residual de cuadrados):



\begin{align}
\large
RSS=\sum_{j=1}^J\sum_{i \in R_j}
\left(
  y_i- \color{red}{ \hat{y}_{R_j}}
  \right)^2
\end{align}

*  $RSS$ es la funcin de prdida que debe minimizarse.

*  $J$ es el nmero total de nodos hoja.

*  El valor pronosticado en la j-sima regin $\color{red}{\hat{y}_{R_j}}$ viene dado por el promedio de esa regin.

*  Para las observaciones pertenecientes a la regin $R_j$, se asigna la misma respuesta predicha $\color{red}{\hat{y}_{R_j}}$ de forma anloga al rbol de clasificacin.

#### $\color{blue}{\text{|}}$ **Criterios de divisin del rbol de decisin**

$\color{blue}{\to}$ El rbol de decisin asigna una pregunta a cada nodo y separa los datos mediante la bifurcacin segn la respuesta.

$\color{blue}{\to}$ Para evaluar qu tan bien se separan los datos, se requiere un criterio especfico. En general, la impureza es el criterio de evaluacin. La impureza aumenta a medida que se mezclan varias clasificaciones en el nodo; Es el ms bajo cuando solo hay una clasificacin.

$\color{blue}{\to}$ Por lo tanto, si la impureza de cada nodo es baja despus de la divisin del nodo, podemos decir que el rbol de decisin est bien clasificado.

$\color{blue}{\to}$ Las medidas de impureza incluyen la impureza y la entropa de Gini.




\begin{align}
\large
\text{Coeficiente Gini}=1-\sum_{i=1}^kp_i^2 \text{   ,
  Coeficiente Entropa}=-\sum_{i=1}^Kp_ilog_2p_i
\end{align}

$\color{blue}{\to}$ El ndice de impurezas es $0$ cuando $p_i = 0$ o $p_i = 1$, y se vuelve ms grande cuando $p_i = \frac{1}{2}$, formando as una parbola.

En otras palabras, el ndice de impurezas es el ms bajo cuando hay una determinada clasificacin en el nodo o el nodo est completamente libre de cualquier clasificacin.

Por el contrario, la impureza se convierte en la ms grande cuando se encuentran muchas clasificaciones en el mismo nodo.

## **Unidad 5. Algoritmo Nave Bayes**

### **| 5.1 Algoritmo Nave Bayes**

$\color{blue}{\text{|}}$ **A cerca del algoritmo Nave Bayes**

$\color{blue}{\to}$ Es una aplicacin directa del teorema de Bayes.

$\color{blue}{\to}$ Nave Bayes no se encuentra entre los algoritmos ms potentes.

Sin embargo, es fcil de entender e implementar.

<br>

$\color{blue}{\text{|}}$ **Pros**
*  Intuitivo y sencillo

*  No es tan sensible al ruido y a los valores atpicos

*  Rpido

<br>

$\color{blue}{\text{|}}$ **Contras**
*  Asume que las caractersticas son independientes, lo que puede no ser estrictamente cierto

*  No se encuentra entre los algoritmos con mejor rendimiento


<br>

$\color{blue}{\to}$ Teorema de Bayes

\begin{align}
\large
P(A|B)=\frac{P(B|A)P(A)}{P(B)}
\end{align}

<br>

$\color{blue}{\to}$ Ahora tomamos A=Clase y B=Datos, entonces:


\begin{align}
\large
P_{post}(Clase)=P(Clase|Dato)=\frac{P(Dato|Clase)P_{prior(Clase)}}{P(Dato)}
\end{align}

<br>

$\color{blue}{\to}$ Estamos ms interesados en comparar probabilidades relativas:



\begin{align}
\large
P_{post}(Clase)\propto P(Dato|Clase)P_{prior}(Clase)
\end{align}

<br>

El Nave Bayes es una aplicacin directa del teorema de Bayes.

En la clasificacin Nave Bayes, nos gustara calcular la probabilidad "posterior" de una clase dados los datos.

Entonces, podremos hacer predicciones comparando las probabilidades de clase.


$\color{blue}{\to}$ Podemos aproximar $P(DatosClase)$ mediante una gaussiana (distribucin normal).

Una gaussiana (distribucin normal) est parametrizada por $\mu$ y $\sigma$.

En este caso, los $\mu_j$ y $\sigma_j$ que corresponden a la clase j se "aprenden" de los datos de entrenamiento.


\begin{align}
\large
P_{post}(Clase)\frac{1}{\sqrt{2\pi}\sigma_j}exp
\left(
  -\frac{1}{2\sigma_j^2}(x-u_j)^2
  \right)P_{prior}(Clase)
\end{align}

donde los parmetros $_j$ y $\sigma_j$ se "aprenden" de los datos de entrenamiento

##### 1. Clasificacin con Naive Bayes:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
%matplotlib inline
```

###### 1.1. Cargar los datos y explorar:

```python
# Cargar los datos.
data = load_iris()
```

```python
# Mostrar la descripcin.
print(data.DESCR)
```

```python
# Variables Explicativas.
X = data['data']
columns = list(data['feature_names'])
print(columns)
```

```python
# Variable de respuesta.
Y = data['target']
labels = list(data['target_names'])
print(labels)
```

```python
# Visualiza la tabla de frecuencias.
ser = pd.Series(Y)
table = ser.value_counts()
table = table.sort_index()                          # Debe ser organizado para etiquetear correctamente
raw_data = {'x': labels, 'y': table.values}
sns.barplot(x='x', y='y', data=raw_data)
plt.show()
```

```python
# Arreglo de scatter plots.
X_df = pd.DataFrame(X,columns=['Sepal_L','Sepal_W','Petal_L','Petal_W'])
my_cols_dict = {0:'red', 1:'green', 2:'blue'}
my_cols = pd.Series(Y).apply(lambda x: my_cols_dict[x])
pd.plotting.scatter_matrix(X_df, c=my_cols, marker='o', alpha=0.5)
plt.show()
```

###### 1.2. Naive Bayes entrenamiento y evaluacin:

```python
# Divide el dataset en evaluacin y prueba.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Entrenar y predecir.
GNB = GaussianNB()
GNB.fit(X_train,Y_train)
Y_pred_test = GNB.predict(X_test)
```

```python
# Matriz de confusin.
conf_mat = metrics.confusion_matrix(Y_test, Y_pred_test)
print(conf_mat)
```

```python
# Accuracy.
accuracy = metrics.accuracy_score(Y_test, Y_pred_test)
print('Accuracy    = ' + str(np.round(accuracy,3)))
```

###### 1.3. Visualiza la probabilidad posterior:

```python
centers = GNB.theta_
variances = GNB.var_
```

```python
for i in range(4):
    x_min = X[:,i].min()
    x_max = X[:,i].max()
    x_range = x_max-x_min
    x_grid = np.linspace(x_min-x_range/3,x_max+x_range/3,300)
    fig=plt.figure(figsize=(4,2), dpi=100)      # Ancho, alto y configuracin DPI
    ax = fig.add_axes([0,0,1,1])                # Izquierda, base, ancho, altura
    ax.set_xlabel(columns[i])
    ax.set_ylabel('Probabilidad')
    for j in range(3):
        center = centers[j,i]
        sigma = np.sqrt(variances[j,i])
        ax.plot(x_grid, st.norm.pdf(x_grid,loc=center,scale=sigma),color=my_cols_dict[j],linestyle="--",label=labels[j])
    ax.legend(loc=0)
    plt.show()
```

## **Unidad 6. Algoritmo KNN**

### **| 5.1 Algoritmo KNN**

<br>

**Acerca de KNN (K-Nearest Neighbors o K-Vecinos ms cercanos)**

$\color{blue}{\to}$ Uno de los algoritmos ms sencillos
La prediccin se basa en los $k$ puntos vecinos ms cercanos.

<br>

$\color{blue}{\to}$ Existen variantes de clasificacin y regresin.

*  Clasificacin: prediccin decidida por la clase mayoritaria de los vecinos ms cercanos.

*  Regresin: prediccin dada por el promedio de los vecinos ms cercanos.


<br>

$\color{blue}{\to}$ KNN es una tcnica de clasificacin y prediccin de un conjunto de datos con la categora desconocida de la variable objetivo. Designa la categora ms similar del conjunto de datos circundante.

$\color{blue}{\to}$ En el caso de KNN, se requieren criterios especficos para medir la "analoga" entre un determinado conjunto de datos y el conjunto de datos circundante y cuntos conjuntos de datos se utilizarn para la clasificacin final de la categora de variable objetivo.

<br>

**1)** Medicin por analoga
Hay muchas maneras de medir la analoga entre los datos. En general, el clculo de analogas se realiza invirtiendo la distancia euclidiana al cuadrado entre dos puntos o utilizando el coeficiente de correlacin de Pearson. Si los puntos son variables discretas, utilice el coeficiente de Jaccard.

<br>

**2)** Criterios de clasificacin de variables objetivas
K en el KNN se refiere al nmero de puntos de datos circundantes que se utilizan para clasificar las variables objetivas del punto de datos especfico despus de medir la analoga entre el punto de datos especfico y otros conjuntos de datos circundantes.


**Ejemplo.**

 *  Tomemos como ejemplo las recomendaciones de pelculas. Recomienda una pelcula que sea similar a la pelcula favorita del cliente y que el cliente an no haya visto. Al considerar solo una pelcula que es la ms similar a la favorita del cliente, es '1-vecino ms cercano'. Al considerar tres pelculas similares, sera '3-vecino ms cercano'.

<br>

 *  Del mismo modo, K-NN es una tcnica para determinar una nueva categora de acuerdo con el principio de la regla de la mayora. Encuentra $k$ puntos de datos circundantes similares al punto de datos especfico y considera la categora de clasificacin del punto de datos especfico.



<br>

La ventaja de un modelo paramtrico es que el patrn de datos se resume en un conjunto relativamente pequeo de parmetros.

Aunque se debe dedicar tiempo y esfuerzo al entrenamiento, la prediccin mediante un modelo previamente entrenado suele ser rpida y eficaz.

KNN es un modelo no paramtrico.
No hay ningn parmetro que se ajuste; No es necesario dedicar tiempo a la formacin.

En cuanto a la prediccin, es bastante ineficiente.



SIC2023_C5_U5_356.png

$\color{blue}{\to}$ En la figura anterior se conceptualiza el cambio en las variables objetivo en funcin de la configuracin del valor K en el mtodo KNN. El punto '$\bigstar$' es el valor de los datos que deben clasificarse; Los puntos en forma de cuadrado y crculo son otros puntos de datos presentes.

<br>

$\color{blue}{\to}$ Al establecer K=1, dado que el punto de datos ms cercano a la estrella es un crculo, la variable objetivo de '$\bigstar$' se clasificar como '$\bigcirc$'. Por otro lado, si K=3, se considerarn los tres puntos ms cercanos a '$\bigstar$', incluyendo dos cuadrados y un crculo. En este caso, '$\bigstar$' se clasificar como '$\square$'. Adems, si K = 5, dado que hay ms crculos alrededor de '$\bigstar$', se clasificar nuevamente como '$\bigcirc$'.

<br>

$\color{blue}{\to}$ Por lo tanto, diferentes valores de K cambian significativamente el resultado predicho del KNN de la categora de variable objetivo. Esta es la razn por la que es importante establecer un valor K apropiado en el mtodo KNN.

<br>

$\color{blue}{\to}$ Sin embargo, no existen estndares tericos o estadsticos claros para un valor K apropiado. Por lo general, se establecen diferentes valores de K para pruebas repetidas, y el valor final de K se establece una vez que muestra el rendimiento ptimo de la clasificacin. Por lo general, se designa un valor K inicial aleatorio entre K = 3 y K = 9, que luego se usa para probar el rendimiento de la clasificacin con datos de entrenamiento y evaluacin para seleccionar el valor K ptimo.

<br>


$\color{blue}{\text{|}}$ **Pros**

$\color{blue}{\to}$ Sencillo e intuitivo

$\color{blue}{\to}$ No hay parmetros de modelo para calcular. Por lo tanto, no hay ningn paso de entrenamiento.

<br>

$\color{blue}{\text{|}}$ **Contras**

$\color{blue}{\to}$ Dado que no existe un "modelo", se puede extraer poca informacin.

$\color{blue}{\to}$ No hay parmetros de modelo que almacenen el patrn aprendido. El conjunto de datos de entrenamiento es necesario para la prediccin.

$\color{blue}{\to}$ La prediccin no es eficiente $\Rightarrow$ "Algoritmo perezoso" o "Lazy Algorithm".

###### 1.5. Entrenamiento y Evaluacin de KNN:

```python
X = df.drop(columns=['Survived'])
Y = df.Survived
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# KNN con n_neighbours = 5
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X_train, Y_train);
Y_pred = knn5.predict(X_test)
print(metrics.confusion_matrix(Y_test,Y_pred))
print("------------------------")
print( "Accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# KNN con n_neighbours = 100
knn100 = KNeighborsClassifier(n_neighbors=100)
knn100.fit(X_train, Y_train);
Y_pred = knn100.predict(X_test)
print(metrics.confusion_matrix(Y_test,Y_pred))
print("------------------------")
print( "Accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

###### 1.6. Compensacin de sesgo-varianza de KNN como funcin de *k*:

```python
accs = []
k_grid = range(1,100,1)
for k in k_grid:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    accs.append(metrics.accuracy_score(Y_test,Y_pred))
```

```python
# Visualiza.
plt.scatter(k_grid,accs,c='red',marker='o',s=10,alpha=0.6)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k')
plt.show()
```

###### 1.7. Optimizacin de hiperparmetros de KNN:

```python
# Malla de parmetros.
k_grid = np.arange(1,51,1)
parameters = {'n_neighbors':k_grid}
```

```python
# Optimiza el k.
gridCV = GridSearchCV(KNeighborsClassifier(), parameters, cv=10, n_jobs = -1)   # "n_jobs = -1" significa "usa todos los CPU cores".
gridCV.fit(X_train, Y_train)
best_k = gridCV.best_params_['n_neighbors']
print("Best k : " + str(best_k))
```

```python
# Prueba con el mejor k.
KNN_best = KNeighborsClassifier(n_neighbors=best_k)
KNN_best.fit(X_train, Y_train)
Y_pred = KNN_best.predict(X_test)
print( "Best Accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```
