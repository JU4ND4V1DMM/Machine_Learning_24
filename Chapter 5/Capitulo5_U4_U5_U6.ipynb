$\color{blue}{\text{|}}$ **Criterio de división (split)**

$\color{blue}{\to}$ Crea el árbol de clasificación de modo que la pureza del nodo secundario sea mayor que la del nodo principal.


SIC2023_C5_U4_319.png

$\color{blue}{\text{|}}$ **Propósito**

$\color{blue}{\to}$ Divide todo el espacio en rectángulos y haz que cada uno sea lo más puro u homogéneo posible.

<br>

$\color{blue}{\to}$ Definición de **«puro»**:
*  Divide el área en espacios rectangulares puros u homogéneos tanto como sea posible

*  Todas las variables del rectángulo final pertenecen al mismo grupo


<br>

En el siguiente gráfico se ilustra un ejemplo relacionando tamaño de vivienda con ingresos y clasificación de propietario o no propietario:




SIC2023_C5_U4_320.png

<br>

$\color{blue}{\text{|}}$ **Proceso de división repetitivo**


**1)** Selecciona $x_i$, una de las variables, y el valor $x_i$ ($s_i$ como criterio de división) se designa para dividir el espacio de dimensión $p$ en dos.

**2)** $x_i \to \{x_i <= s_i\} \cup \{x_i > s_i\}$

**3)** Selecciona una variable de nuevo y divídela de la misma manera.

**4)** Repite el proceso hasta que alcance la pureza deseada.



<br>

#### $\color{blue}{\text{|}}$ **Criterio de División**

$\color{blue}{\to}$ Estadístico Chi-cuadrado - p-value: crea nodos secundarios con una variable predictora con el p-value más bajo y la partición óptima

$\color{blue}{\to}$ Índice de Gini: selecciona nodos secundarios con una variable predictora que reduce el índice de Gini y la partición óptima

$\color{blue}{\to}$ Medida de entropía: crea nodos secundarios con una variable predictora con la medida de entropía más baja y la partición óptima


<br>

Vs. coeficiente de Gini en economía: 0 (igualdad) ~ 1 (desigualdad)

Entropía: Grado de incertidumbre
La entropía se puede definir tanto termodinámica como estadísticamente.

La entropía es una función de estado utilizada para explicar el flujo de energía termodinámicamente no útil (energía que no se puede convertir en trabajo).

Estadísticamente, la entropía es un logaritmo del número de estado microscópico que responde al estado macroscópico. En general, la entropía no se conserva. Aumentó a través del tiempo de acuerdo con la segunda ley de la termodinámica.


<br>

#### $\color{blue}{\text{|}}$ **Variables objetivo continuas**


**Estadístico F en ANOVA**: crea nodos secundarios con una variable predictora con el valor $p$ más bajo y la partición óptima

**Reducción de la varianza**: crea nodos secundarios con la creación de particiones óptima que maximiza la reducción de la varianza

#### $\color{blue}{\text{|}}$ **Medida de Impureza**

$\color{blue}{\to}$ Selecciona los nodos secundarios con una variable predictora que reduce el índice de Gini y la creación de particiones óptima.

$\color{blue}{\to}$ Si el conjunto de datos $T$ se divide en $k$ categorías y los coeficientes de rendimiento de la categoría son $p_1,\cdots, p_k$, se expresa como la siguiente ecuación.


\begin{align}
\large
Gini(T)=1-\sum_{l=1}^kp_l^2
\end{align}

SIC2023_C5_U4_324.png

#### $\color{blue}{\text{|}}$ **Medición de Entropía**

$\color{blue}{\to}$ En termodinámica, la entropía mide el grado de desorden.

$\color{blue}{\to}$ Crea nodos secundarios con una variable predictora con la medida de entropía más baja y la creación de particiones óptima.

$\color{blue}{\to}$ Si el conjunto de datos $T$ se divide en $k$ categorías y los coeficientes de rendimiento de la categoría son $p_1, \cdots, p_k$, se expresa como la siguiente ecuación.


\begin{align}
\large
Entropía(T)=-\sum_{l=1}^kp_llog_2p_l
\end{align}


<br>

**Ejemplo.** Si 4 categorías consisten en proporciones de 0,25, 0,25, 0,25, 0,25 ($T_0$):

\begin{align}
\large
Entropía(T_0)=-(0.25log_20.25) \times 4 = 2
\end{align}

<br>

**Ejemplo.** Si 4 categorías consisten en proporciones de 0.5, 0.25, 0.25, 0 ($T_1$):

\begin{align}
\large
Entropía(T_1)=-(0.5log_20.5+0.25log_20.25+0.25log_20.25)=1.5
\end{align}

#### $\color{blue}{\text{|}}$ **Árbol de clasificación: Procedimiento**

**a)** Hacer un árbol básico.

**b)** Podar las ramas que no proporcionen un mejor rendimiento. La poda se puede realizar durante el paso de validación cruzada.

**c)** Predecir con el árbol optimizado.

La poda se lleva a cabo poniendo restricciones al crecimiento de la estructura del árbol.

#### $\color{blue}{\text{|}}$ **Arbol de Regresión**

$\color{blue}{\to}$ El espacio configuracional se divide en regiones: $\{R_1,R_2, ...,R_J\}$.

$\color{blue}{\to}$ La estructura de árbol se entrena minimizando el RSS (suma residual de cuadrados):



\begin{align}
\large
RSS=\sum_{j=1}^J\sum_{i \in R_j}
\left(
  y_i- \color{red}{ \hat{y}_{R_j}}
  \right)^2
\end{align}

*  $RSS$ es la función de pérdida que debe minimizarse.

*  $J$ es el número total de nodos hoja.

*  El valor pronosticado en la j-ésima región $\color{red}{\hat{y}_{R_j}}$ viene dado por el promedio de esa región.

*  Para las observaciones pertenecientes a la región $R_j$, se asigna la misma respuesta predicha $\color{red}{\hat{y}_{R_j}}$ de forma análoga al árbol de clasificación.

#### $\color{blue}{\text{|}}$ **Árbol de Decisión**

$\color{blue}{\to}$ El árbol de decisión se refiere a una técnica de modelado como la forma de un árbol que se ramifica desde la raíz hasta los nodos de la hoja.

<br>

$\color{blue}{\to}$ Depende de los valores de referencia de las variables independientes (variables explicativas o de entrada) que afectan a la clasificación o predicción de las variables objetivo.

<br>

$\color{blue}{\to}$ En el árbol de decisión, cada nodo se divide en forma de **si-entonces** en función de las características de las variables explicativas o de los valores de referencia.
Al seguir la estructura de árbol, es posible comprender fácilmente cómo se clasifica el valor del atributo de los datos en la categoría.

#### $\color{blue}{\text{|}}$ **Criterios de división del árbol de decisión**

$\color{blue}{\to}$ El árbol de decisión asigna una pregunta a cada nodo y separa los datos mediante la bifurcación según la respuesta.

$\color{blue}{\to}$ Para evaluar qué tan bien se separan los datos, se requiere un criterio específico. En general, la impureza es el criterio de evaluación. La impureza aumenta a medida que se mezclan varias clasificaciones en el nodo; Es el más bajo cuando solo hay una clasificación.

$\color{blue}{\to}$ Por lo tanto, si la impureza de cada nodo es baja después de la división del nodo, podemos decir que el árbol de decisión está bien clasificado.

$\color{blue}{\to}$ Las medidas de impureza incluyen la impureza y la entropía de Gini.




\begin{align}
\large
\text{Coeficiente Gini}=1-\sum_{i=1}^kp_i^2 \text{   ,
  Coeficiente Entropía}=-\sum_{i=1}^Kp_ilog_2p_i
\end{align}

$\color{blue}{\to}$ El índice de impurezas es $0$ cuando $p_i = 0$ o $p_i = 1$, y se vuelve más grande cuando $p_i = \frac{1}{2}$, formando así una parábola.

En otras palabras, el índice de impurezas es el más bajo cuando hay una determinada clasificación en el nodo o el nodo está completamente libre de cualquier clasificación.

Por el contrario, la impureza se convierte en la más grande cuando se encuentran muchas clasificaciones en el mismo nodo.

###### 1.2. Optimización de Hiperparámetros del Árbol:

```python
depth_grid = np.arange(1,21)
min_samples_leaf_grid = np.arange(10,31)
max_leaf_nodes_grid = np.arange(2,21)
parameters = {'max_depth':depth_grid, 'min_samples_leaf':min_samples_leaf_grid, 'max_leaf_nodes':max_leaf_nodes_grid}
```

```python
print("Mejor depth (profundidad) del árbol : " + str(best_depth))
print("Mejor min_samples_leaf del árbol : " + str(best_min_samples_leaf))
print("Mejor max_leaf_nodes del árbol : " + str(best_max_leaf_nodes))
```

```python
DTC_best = DecisionTreeClassifier(max_depth=best_depth,min_samples_leaf=best_min_samples_leaf,max_leaf_nodes=best_max_leaf_nodes)
DTC_best.fit(X_train, Y_train);
Y_pred = DTC_best.predict(X_test)
print( "Mejor presición del árbol : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
print( "Mejor RMSE del Árbol : " + str(np.round(np.sqrt(metrics.mean_squared_error(Y_test,Y_pred)),3)))
```

```python
print( "Mejor presición del árbol : " + str(np.round(metrics.recall_score(Y_test,Y_pred),3)))
```

## **Unidad 5. Algoritmo Naïve Bayes**

### **| 5.1 Algoritmo Naïve Bayes**

$\color{blue}{\text{|}}$ **A cerca del algoritmo Naïve Bayes**

$\color{blue}{\to}$ Es una aplicación directa del teorema de Bayes.

$\color{blue}{\to}$ Naïve Bayes no se encuentra entre los algoritmos más potentes.

Sin embargo, es fácil de entender e implementar.

<br>

$\color{blue}{\text{|}}$ **Pros**
*  Intuitivo y sencillo

*  No es tan sensible al ruido y a los valores atípicos

*  Rápido

<br>

$\color{blue}{\text{|}}$ **Contras**
*  Asume que las características son independientes, lo que puede no ser estrictamente cierto

*  No se encuentra entre los algoritmos con mejor rendimiento


<br>

$\color{blue}{\to}$ Teorema de Bayes

\begin{align}
\large
P(A|B)=\frac{P(B|A)P(A)}{P(B)}
\end{align}

<br>

$\color{blue}{\to}$ Ahora tomamos A=Clase y B=Datos, entonces:


\begin{align}
\large
P_{post}(Clase)=P(Clase|Dato)=\frac{P(Dato|Clase)P_{prior(Clase)}}{P(Dato)}
\end{align}

<br>

$\color{blue}{\to}$ Estamos más interesados en comparar probabilidades relativas:



\begin{align}
\large
P_{post}(Clase)\propto P(Dato|Clase)P_{prior}(Clase)
\end{align}

<br>

El Naïve Bayes es una aplicación directa del teorema de Bayes.

En la clasificación Naïve Bayes, nos gustaría calcular la probabilidad "posterior" de una clase dados los datos.

Entonces, podremos hacer predicciones comparando las probabilidades de clase.


$\color{blue}{\to}$ Podemos aproximar $P(Datos│Clase)$ mediante una gaussiana (distribución normal).

Una gaussiana (distribución normal) está parametrizada por $\mu$ y $\sigma$.

En este caso, los $\mu_j$ y $\sigma_j$ que corresponden a la clase j se "aprenden" de los datos de entrenamiento.


\begin{align}
\large
P_{post}(Clase)∝\frac{1}{\sqrt{2\pi}\sigma_j}exp
\left(
  -\frac{1}{2\sigma_j^2}(x-u_j)^2
  \right)P_{prior}(Clase)
\end{align}

donde los parámetros $μ_j$ y $\sigma_j$ se "aprenden" de los datos de entrenamiento

##### 1. Clasificación con Naive Bayes:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
%matplotlib inline
```

###### 1.1. Cargar los datos y explorar:

```python
# Cargar los datos.
data = load_iris()
```

```python
# Mostrar la descripción.
print(data.DESCR)
```

```python
# Variables Explicativas.
X = data['data']
columns = list(data['feature_names'])
print(columns)
```

```python
# Variable de respuesta.
Y = data['target']
labels = list(data['target_names'])
print(labels)
```

```python
# Visualiza la tabla de frecuencias.
ser = pd.Series(Y)
table = ser.value_counts()
table = table.sort_index()                          # Debe ser organizado para etiquetear correctamente
raw_data = {'x': labels, 'y': table.values}
sns.barplot(x='x', y='y', data=raw_data)
plt.show()
```

```python
# Arreglo de scatter plots.
X_df = pd.DataFrame(X,columns=['Sepal_L','Sepal_W','Petal_L','Petal_W'])
my_cols_dict = {0:'red', 1:'green', 2:'blue'}
my_cols = pd.Series(Y).apply(lambda x: my_cols_dict[x])
pd.plotting.scatter_matrix(X_df, c=my_cols, marker='o', alpha=0.5)
plt.show()
```

###### 1.2. Naive Bayes entrenamiento y evaluación:

```python
# Divide el dataset en evaluación y prueba.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Entrenar y predecir.
GNB = GaussianNB()
GNB.fit(X_train,Y_train)
Y_pred_test = GNB.predict(X_test)
```

```python
# Matriz de confusión.
conf_mat = metrics.confusion_matrix(Y_test, Y_pred_test)
print(conf_mat)
```

```python
# Accuracy.
accuracy = metrics.accuracy_score(Y_test, Y_pred_test)
print('Accuracy    = ' + str(np.round(accuracy,3)))
```

###### 1.3. Visualiza la probabilidad posterior:

```python
centers = GNB.theta_
variances = GNB.var_
```

```python
for i in range(4):
    x_min = X[:,i].min()
    x_max = X[:,i].max()
    x_range = x_max-x_min
    x_grid = np.linspace(x_min-x_range/3,x_max+x_range/3,300)
    fig=plt.figure(figsize=(4,2), dpi=100)      # Ancho, alto y configuración DPI
    ax = fig.add_axes([0,0,1,1])                # Izquierda, base, ancho, altura
    ax.set_xlabel(columns[i])
    ax.set_ylabel('Probabilidad')
    for j in range(3):
        center = centers[j,i]
        sigma = np.sqrt(variances[j,i])
        ax.plot(x_grid, st.norm.pdf(x_grid,loc=center,scale=sigma),color=my_cols_dict[j],linestyle="--",label=labels[j])
    ax.legend(loc=0)
    plt.show()
```

## **Unidad 6. Algoritmo KNN**

### **| 5.1 Algoritmo KNN**

<br>

**Acerca de KNN (K-Nearest Neighbors o K-Vecinos más cercanos)**

$\color{blue}{\to}$ Uno de los algoritmos más sencillos
La predicción se basa en los $k$ puntos vecinos más cercanos.

<br>

$\color{blue}{\to}$ Existen variantes de clasificación y regresión.

*  Clasificación: predicción decidida por la clase mayoritaria de los vecinos más cercanos.

*  Regresión: predicción dada por el promedio de los vecinos más cercanos.


<br>

$\color{blue}{\to}$ KNN es una técnica de clasificación y predicción de un conjunto de datos con la categoría desconocida de la variable objetivo. Designa la categoría más similar del conjunto de datos circundante.

$\color{blue}{\to}$ En el caso de KNN, se requieren criterios específicos para medir la "analogía" entre un determinado conjunto de datos y el conjunto de datos circundante y cuántos conjuntos de datos se utilizarán para la clasificación final de la categoría de variable objetivo.

<br>

**1)** Medición por analogía
Hay muchas maneras de medir la analogía entre los datos. En general, el cálculo de analogías se realiza invirtiendo la distancia euclidiana al cuadrado entre dos puntos o utilizando el coeficiente de correlación de Pearson. Si los puntos son variables discretas, utilice el coeficiente de Jaccard.

<br>

**2)** Criterios de clasificación de variables objetivas
K en el KNN se refiere al número de puntos de datos circundantes que se utilizan para clasificar las variables objetivas del punto de datos específico después de medir la analogía entre el punto de datos específico y otros conjuntos de datos circundantes.


**Ejemplo.**

 *  Tomemos como ejemplo las recomendaciones de películas. Recomienda una película que sea similar a la película favorita del cliente y que el cliente aún no haya visto. Al considerar solo una película que es la más similar a la favorita del cliente, es '1-vecino más cercano'. Al considerar tres películas similares, sería '3-vecino más cercano'.

<br>

 *  Del mismo modo, K-NN es una técnica para determinar una nueva categoría de acuerdo con el principio de la regla de la mayoría. Encuentra $k$ puntos de datos circundantes similares al punto de datos específico y considera la categoría de clasificación del punto de datos específico.



<br>

La ventaja de un modelo paramétrico es que el patrón de datos se resume en un conjunto relativamente pequeño de parámetros.

Aunque se debe dedicar tiempo y esfuerzo al entrenamiento, la predicción mediante un modelo previamente entrenado suele ser rápida y eficaz.

KNN es un modelo no paramétrico.
No hay ningún parámetro que se ajuste; No es necesario dedicar tiempo a la formación.

En cuanto a la predicción, es bastante ineficiente.



SIC2023_C5_U5_356.png

$\color{blue}{\to}$ En la figura anterior se conceptualiza el cambio en las variables objetivo en función de la configuración del valor K en el método KNN. El punto '$\bigstar$' es el valor de los datos que deben clasificarse; Los puntos en forma de cuadrado y círculo son otros puntos de datos presentes.

<br>

$\color{blue}{\to}$ Al establecer K=1, dado que el punto de datos más cercano a la estrella es un círculo, la variable objetivo de '$\bigstar$' se clasificará como '$\bigcirc$'. Por otro lado, si K=3, se considerarán los tres puntos más cercanos a '$\bigstar$', incluyendo dos cuadrados y un círculo. En este caso, '$\bigstar$' se clasificará como '$\square$'. Además, si K = 5, dado que hay más círculos alrededor de '$\bigstar$', se clasificará nuevamente como '$\bigcirc$'.

<br>

$\color{blue}{\to}$ Por lo tanto, diferentes valores de K cambian significativamente el resultado predicho del KNN de la categoría de variable objetivo. Esta es la razón por la que es importante establecer un valor K apropiado en el método KNN.

<br>

$\color{blue}{\to}$ Sin embargo, no existen estándares teóricos o estadísticos claros para un valor K apropiado. Por lo general, se establecen diferentes valores de K para pruebas repetidas, y el valor final de K se establece una vez que muestra el rendimiento óptimo de la clasificación. Por lo general, se designa un valor K inicial aleatorio entre K = 3 y K = 9, que luego se usa para probar el rendimiento de la clasificación con datos de entrenamiento y evaluación para seleccionar el valor K óptimo.

<br>


$\color{blue}{\text{|}}$ **Pros**

$\color{blue}{\to}$ Sencillo e intuitivo

$\color{blue}{\to}$ No hay parámetros de modelo para calcular. Por lo tanto, no hay ningún paso de entrenamiento.

<br>

$\color{blue}{\text{|}}$ **Contras**

$\color{blue}{\to}$ Dado que no existe un "modelo", se puede extraer poca información.

$\color{blue}{\to}$ No hay parámetros de modelo que almacenen el patrón aprendido. El conjunto de datos de entrenamiento es necesario para la predicción.

$\color{blue}{\to}$ La predicción no es eficiente $\Rightarrow$ "Algoritmo perezoso" o "Lazy Algorithm".

###### 1.6. Compensación de sesgo-varianza de KNN como función de *k*:

```python
accs = []
k_grid = range(1,100,1)
for k in k_grid:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    accs.append(metrics.accuracy_score(Y_test,Y_pred))
```
