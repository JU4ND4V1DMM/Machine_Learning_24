### **| 7.1 Algoritmo SVM**

$\color{blue}{\text{|}}$ **A cerca de SVM (Mquina de Soporte Vectorial o Supported Vector Machine)**


$\color{blue}{\to}$ Mejora de la precisin de la clasificacin al maximizar el margen

$\color{blue}{\to}$ Lmite de clasificacin no lineal efectivo por la transformacin "kernel"

<br>

SVM introdujo muchos conceptos nuevos.

Se trata de un algoritmo muy eficaz que puede funcionar bien incluso cuando se requiere un lmite de clasificacin no lineal.




<br>

$\color{blue}{\to}$ El SVM es un modelo que clasifica los datos encontrando la lnea (o hiperplano) donde la distancia (margen) entre los datos en diferentes categoras se vuelve mxima.

<br>

$\color{blue}{\to}$ Como se muestra en la figura siguiente, el modelo de SVM encuentra el hiperplano que divide los datos en dos categoras diferentes con el margen mximo para clasificar los datos.  

<br>

$\color{blue}{\to}$ Habra muchas lneas o planos que dividen los datos en dos categoras diferentes. Sin embargo, los puntos que superan el lmite de clasificacin pueden aparecer en datos de evaluacin o datos futuros desconocidos a medida que los puntos cercanos al lmite cambian ligeramente.


<br>

 SVM es un algoritmo potente. Sin embargo, se necesita ms tiempo y esfuerzo para entrenarlo adecuadamente.


<br>

$\color{blue}{\to}$ Para minimizar esta posibilidad, se debe encontrar una lnea (o hiperplano) que haga el mximo margen entre las diferentes categoras de datos. En otras palabras, el objetivo es encontrar un hiperplano que pueda inducir la mxima diferenciacin para la mejor generalizacin posible para clasificar y predecir los datos futuros, no los datos de entrenamiento actuales.

<br>

$\color{blue}{\to}$ El punto de la categora ms cercano a la lnea lmite se denomina vector de soporte, y cada clasificacin debe tener al menos un vector de soporte.

<br>


<br>

$\color{blue}{\to}$ Sin embargo, no siempre es posible clasificar todos los datos de forma lineal. A veces, la clasificacin de datos debe realizarse en una curva o en un plano de clasificacin no lineal ms complejo.

<br>

$\color{blue}{\to}$ Si es as, utiliza el mtodo de truco Kernel para asignar los datos especificados a una dimensin superior y busque un hiperplano que pueda clasificar los datos en la dimensin convertida.

<br>

$\color{blue}{\to}$ En lugar de convertir los datos a una dimensin superior, el mtodo de truco Kernel utiliza la funcin Kernel que devuelve un valor similar al que se realiza cuando se realizan clculos internos entre vectores en la dimensin superior. Esto puede dar lugar a una clasificacin no lineal sin mover los datos a una dimensin superior.

<br>

$\color{blue}{\to}$ Las principales funciones del kernel incluyen ncleos polinmicos, gaussianos y sigmoides.




<br>

$\color{blue}{\text{|}}$ **Pros**

$\color{blue}{\to}$ No es muy sensible a los valores atpicos.

$\color{blue}{\to}$ El rendimiento es bueno.

<br>

$\color{blue}{\text{|}}$ **Contras**

$\color{blue}{\to}$ El entrenamiento es relativamente lento. Funciona mal para datos de gran tamao.

$\color{blue}{\to}$ El kernel y el conjunto de hiperparmetros deben optimizarse cuidadosamente.

$\color{blue}{\to}$ No se puede obtener mucha informacin.

$\color{blue}{\text{|}}$ **Hiperplano**

$\color{blue}{\to}$ Para un espacio configuracional k dimensional, el hiperplano tiene la dimensin k1.

$\color{blue}{\to}$ Aqu, expresamos matemticamente el "hiperplano" que divide en dos el espacio configuracional.

<br>

**Ejemplo.**

Para un espacio bidimensional, un hiperplano es una lnea bisectrizable que se puede parametrizar como:


\begin{align}
\large
\beta_0+\beta_1X_1+\beta_2X_2=0
\end{align}

<br>

El espacio bidimensional se subdivide en dos:

\begin{align}
\large
\beta_0+\beta_1X_1+\beta_2X_2>0 \\
\large
\beta_0+\beta_1X_1+\beta_2X_2<0
\end{align}

<br>

Una observacin pertenecera a cualquiera de ellos $\Rightarrow$ clasificacin binaria!


<br>

$\color{blue}{\text{|}}$ **Optimizacin de mrgenes (para el binario $y$)**

SVM consiste en optimizar los parmetros del hiperplano de manera que se maximice el margen.
Al ajustar el hiperparmetro $C$, podemos controlar la "flexibilidad".

Aunque la explicacin se limita a la respuesta binaria ($y$), SVM se puede generalizar a la prediccin multiclase utilizando el esquema "uno contra todos".

<br>

$\color{blue}{\to}$ El margen de lmite se maximiza.

$\color{blue}{\to}$ Si no es posible establecer un margen claro, se permite un error dentro de un lmite.

$\color{blue}{\to}$ El objetivo es maximizar $M$ optimizando los parmetros $\beta_0,\beta_1,\cdots,\beta_k$ sujetos a las restricciones:

<br>

*  **Restriccin 1:** $\sum_{j=1}^k \beta_j^2 =1$

<br>

*  **Restriccin 2:** $y_i (\beta_0+\beta_1 x_{i1}+\beta_2 x_{i2}+...+\beta_k x_{ik} )M\color{red}{(1-\varepsilon_i)}, i=1,2,\cdots,n \\
\\
y_i \text{ es -1 o +1.}$



<br>

*  **Restriccin 3:** $\color{red}{\varepsilon_i \ge 0}$ y $\color{red}{\sum_{i=1}^k\varepsilon_i \le C}$ $\color{green}{\text{ } \Leftarrow}$ $C$ *es un hiperparmetro relacionado con los errores de clasificacin*


<br>

<br>

$\color{blue}{\to}$ Mapeo efectivo a una dimensin superior al dar el producto interno de dos vectores $x_1$ y $x_2$.

<br>

La funcin kernel da el producto interno en dimensiones ms altas.

Hay varios tipos diferentes de funciones del kernel.



<br>

*  Lineal: $K(x_1,x_2 )=x_1^t x_2$

*  Polinomio: $K(x_1,x_2 )=(x_1^t x_2+c)^d$, donde $c>0$

*  Sigmoide: $K(x_1,x_2 )=tanh(a(x_1^t x_2 )+b)$, donde $a, b>0$

*  Base de la funcin radial (rbf): $K(x_1,x_2 )=exp(-\gammax_1-x_2 ^2 )$, donde $\gamma>0$

##### 1. Clasificacin con SVM:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import recall_score
from sklearn.datasets import load_iris
warnings.filterwarnings(action='ignore')                  # Desactivar las advertencias del sistema.
%matplotlib inline
```

###### 1.1. Leer los datos:

```python
# Cargar los datos.
data = load_iris()
```

```python
# Variables explicativas.
X = data['data']
columns = list(data['feature_names'])
print(columns)
```

```python
# Variable de respuesta.
Y = data['target']
labels = list(data['target_names'])
print(labels)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

###### 1.2. Optimizacin de hiperparmetros SVM (RBF kernel):

C     : Parmetro de penalidad.
<br>

gamma : parmetro kernel ($\gamma$).

```python
C_grid = 0.02*np.arange(20,50)
gamma_grid = 0.02*np.arange(1,50)
parameters = {'C': C_grid, 'gamma' : gamma_grid}
gridCV = GridSearchCV(SVC(kernel='rbf'), parameters, cv=10, n_jobs=-1)          # "n_jobs = -1" significa "use todos los CPU cores".
gridCV.fit(X_train, Y_train)
best_C = gridCV.best_params_['C']
best_gamma = gridCV.best_params_['gamma']
```

```python
print("SVM mejor C : " + str(best_C))
print("SVM mejor gamma : " + str(best_gamma))
```

```python
SVM_best = SVC(kernel='rbf', C=best_C, gamma=best_gamma)
SVM_best.fit(X_train, Y_train);
Y_pred = SVM_best.predict(X_test)
print( "SVM mejor accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
print( "SVM mejor recall : " + str(np.round(metrics.recall_score(Y_test,Y_pred, average='weighted'),3)))
```

###### 1.3. Optimizacin de hiperparmetro SVM (kernel polinomial):

```python
C_grid = 0.0001*np.arange(1,100)
gamma_grid = 0.01*np.arange(1,100)
parameters = {'C': C_grid, 'gamma' : gamma_grid}
gridCV = GridSearchCV(SVC(kernel='poly'), parameters, cv=10, n_jobs=-1)         # "n_jobs = -1" significa "use todos los CPU cores".
gridCV.fit(X_train, Y_train)
best_C = gridCV.best_params_['C']
best_gamma = gridCV.best_params_['gamma']
```

```python
print("SVM mejor C : " + str(best_C))
print("SVM mejor gamma : " + str(best_gamma))
```

```python
SVM_best = SVC(kernel='poly', C=best_C,gamma=best_gamma)
SVM_best.fit(X_train, Y_train);
Y_pred = SVM_best.predict(X_test)
print( "SVM mejor accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

$\color{blue}{\text{|}}$ **Qu es aprendizaje Ensemble?**

<br>

$\color{blue}{\to}$ El trmino *Ensemble* puede definirse de la siguiente manera:
*  En mecnica estadstica, un *ensemble* de un sistema se refiere a la coleccin de sistemas equivalentes.

<br>

$\color{blue}{\to}$ Es decir, es un conjunto de grupos similares.

<br>

$\color{blue}{\to}$ En lugar de esperar resultados de rendimiento de un solo modelo, el aprendizaje  ensemble obtiene un mejor resultado utilizando la inteligencia colectiva de diferentes modelos, como promediar muchos modelos individuales diferentes o tomar una decisin basada en el voto de la mayora.

<br>

$\color{blue}{\to}$ Hay muchos mtodos de ensamble diferentes que utilizan la inteligencia colectiva.

*  Votacin: obtencin de resultados a travs de la votacin
*  Embolsado: agregacin de Bootstrap (creacin duplicada de varias muestras)
*  Boosting - Ponderacin complementando errores anteriores
*  Apilamiento: un metamodelo basado en diferentes modelos

<br>

$\color{blue}{\to}$ Puede haber otros mtodos diferentes ms, ya que el aprendizaje por ensemble aplica una determinada tcnica/metodologa. Sin embargo, los cuatro mtodos enumerados anteriormente son las tcnicas de ensemble ms representativas de la librera sklearn.

$\color{blue}{\text{|}}$ **Ensemble de votacin**

$\color{blue}{\to}$ Se puede aplicar a la clasificacin y la regresin.

<br>

$\color{blue}{\to}$ Los modelos que forman un ensemble deben ser de diferentes tipos para un buen desempeo.

<br>

$\color{blue}{\to}$ Dos mtodos de votacin para el clasificador:
*  **Duro**: la etiqueta de clase predicha viene dada por la votacin de la regla de la mayora.

*  **Suave**: la etiqueta de clase predicha viene dada por el $\color{blue}{\text{argmax}}$ de la suma de las probabilidades predichas.

<br>

Podemos hacer un ensemble de votacin en el que los modelos bsicos deben ser de diferentes tipos para un buen desempeo.

El *argmax* significa la posicin en el valor ms grande.

$\color{blue}{\text{|}}$ **Votacin**

$\color{blue}{\to}$ Es bastante fcil implementar un ensemble de votacin con la librera scikit-learn.

<br>

$\color{blue}{\to}$ Como la palabra misma, votacin consiste en tomar una decisin a travs de los votos. La votacin es similar al (bagging) embolsado, ya que utiliza un mtodo de votacin, pero se diferencian entre s de la siguiente manera:

<br>

*  **Votacin**: Combina diferentes modelos de algoritmos.
*  **Embolsado (bagging)**: Utiliza diferentes combinaciones de muestras dentro del mismo algoritmo.

<br>

$\color{blue}{\to}$ La votacin selecciona los resultados finales haciendo que la votacin final sobre los resultados sea deducida por diferentes algoritmos.

<br>

$\color{blue}{\to}$ La votacin se clasifica en voto duro y voto blando.

<br>

*  **Votacin dura**: Decide el valor final del resultado a travs de la votacin.

*  **Votacin suave**: Extrae el valor final sumando todos los valores de probabilidad de obtener el resultado final y luego calculando cada probabilidad del resultado final.

<br>

$\color{blue}{\text{|}}$ **Partes del sistema de votacin**

SIC2023_C5_U8_384.png


$\color{blue}{\to}$ **Votacin dura**

*  Tomando como ejemplo la clasificacin. Supongamos que los valores predictivos arrojados por los **modelos individuales** para la clasificacin son 1,0,0,1,1, ya que la $\color{blue}{\text{clase 1}}$ tiene tres votos y la $\color{red}{\text{clase 0}}$ tiene 2 votos. En ese caso, 1 se convierte en el valor predictivo final en el mtodo de votacin dura.

<br>

$\color{blue}{\to}$ **Votacin blanda**

*  El mtodo de votacin blanda calcula el valor promedio de cada probabilidad y luego determina la que tiene la probabilidad ms alta.

*  Supongamos que los **modelos individuales** arrojan la probabilidad de obtener la $\color{red}{\text{clase 0}}$ es $\color{red}{(0.4, 0.9, 0.9, 0.4, 0.4)}$

  y la probabilidad de obtener la $\color{blue}{\text{clase 1}}$ es $\color{blue}{(0.6, 0.1, 0.1, 0.6, 0.6)}$.

<br>

*  En base a la **votacin blanda** La probabilidad final de obtener la $\color{red}{\text{clase 0}}$ es
$\large\frac{(0,4 + 0,9 + 0,9 + 0,4 + 0,4)}{5} = 0,6$;

  La probabilidad final de obtener la $\color{blue}{\text{clase 1}}$ es
  $\large\frac{$(0,6 + 0,1 + 0,1 + 0,6 + 0,6)}{5} = 0,4$.
  
  Por lo tanto, el valor final seleccionado es diferente del resultado de la votacin dura anterior.

<br>

$\color{blue}{\to}$ En general, el uso del mtodo de votacin blanda se considera ms razonable que el mtodo de votacin dura en las competiciones, ya que el mtodo de votacin blanda proporciona un resultado de rendimiento real mucho mejor.

$\color{blue}{\text{|}}$ **Conjunto de impulso: AdaBoost**

**Acerca de AdaBoost**

$\color{blue}{\to}$ Una secuencia de modelos dbiles, como los rboles

<br>

$\color{blue}{\to}$ Un conjunto ponderado de modelos dbiles

<br>

$\color{blue}{\to}$ Un conjunto de ponderaciones que aumentan la importancia de los modelos con mejor rendimiento

<br>

$\color{blue}{\to}$ Un conjunto de ponderaciones que aumentan la importancia de las observaciones errneamente clasificadas

<br>

$\color{blue}{\to}$ Pros y contras similares a los del Random Forest

<br>

$\color{blue}{\to}$ AdaBoost es un conjunto de modelos dbiles (generalmente rboles).

<br>

$\color{blue}{\to}$ A diferencia de Random Forest, donde los modelos bsicos se entrenan a la vez en paralelo, AdaBoost tiene una secuencia de modelos bsicos que se entrenan uno tras otro.

<br>


$\color{blue}{\text{|}}$ **Algoritmo de Clasificacin AdaBoost**


$\color{blue}{\to}$ Supongamos:

*  $n$ observaciones para el paso de entrenamiento: $x_i$ y $y_i$.

*  Tambin suponemos que $y_i \in \{1,+1\}$.  ($y$ binario)

*  La prediccin se basa en el uso de la funcin $sign()$ que devuelve +1 o 1.




<br>

$\color{blue}{\to}$ Haremos una serie de modelos dbiles $G_m (x)$ con $m= 1,\cdots,M$.

<br>

$\color{blue}{\to}$ El clasificador *ensemble* se compone de una combinacin lineal de estos modelos dbiles:


\begin{align}
\large
G_{ensemble}(x)=sign
\left(  \sum_{m=1}^M\alpha_mG_m(x)
  \right)
\end{align}

<br>

donde $\alpha_m$ son los "pesos de aumento" o "boosting weights" que deben calcularse.


$\color{blue}{\to}$ Le asignamos un mayor peso a un modelo de mejor rendimiento.

<br>

El ensemble en AdaBoost es, en realidad, una secuencia.



<br>

$\color{blue}{\to}$ Cada paso se compone de entrenamiento y prediccin.

<br>

$\color{blue}{\to}$ Se entrena a un modelo bsico y, a continuacin, se estiman los coeficientes de error de la prediccin.


**1)** Para el primer paso ($m=1$), se asigna el mismo peso a las observaciones:

\begin{align}
w_i^{(1)}\frac{-1}{n}
\end{align}

<br>

**2)** Para la secuencia de boos $m = 1,\cdots, M$:

  **a)** Entrena al modelo $G_m (x)$ utilizando observaciones ponderadas por $w_i^{(m)}$.  

  **b)** Calcula la tasa de error:

\begin{align}
\epsilon_m=\frac{\sum_{i=1}^nw_i^{(m)}\color{red}{I(y_i \ne G_m(x_i))}}{\sum_{i=1}^nw_i^{(m)}}
\end{align}

<br>

$\Rightarrow \color{red}{{I(y_i \ne G_m(x_i))}}$ resulta 1 para una prediccin incorrecta, sino resulta 0.

$\Rightarrow 0 \le \epsilon_m \le 1$



<br>

**3)** Para la secuencia de refuerzo $m = 1,\cdots, M$:

**c)** Calcule el peso de *boosting*:

\begin{align}
\alpha_m=\frac{1}{2}log\left(\frac{1-\epsilon_m}{\epsilon_m}\right)
\end{align}

<br>

$\Rightarrow$ Como $\epsilon_m\to0$, $\alpha_m$ es un nmero positivo grande. Al modelo se le da ms importancia!

$\Rightarrow$  Como $\epsilon_m\cong0.5$, $\alpha_m\cong0$.

$\Rightarrow$  Como $\epsilon_m\to1$, $\alpha_m$ es un nmero negativo grande.


<br>

d) Para el siguiente paso, los pesos de la observacin predicha $\color{red}{\text{errneamente}}$ se reescalan mediante un factor $e_m^a$.  

Esto se puede expresar de forma compacta como:

\begin{align}
w_i^{(m+1)}=w_i^{(m)}\times e^{\alpha \cdot m \cdot I(y_i \ne G_m(x_i))}\text{, donde }i=1,\cdots,n
\end{align}

<br>

$\Rightarrow$ En el siguiente paso de la secuencia, las observaciones predichas $\color{red}{\text{errneamente}}$ reciben un mayor peso.

<br>



**4)** El clasificador ensemble se compone de una combinacin lineal de modelos dbiles:

\begin{align}
\large
G_{ensemble}(x)=sign
\left(  \sum_{m=1}^M\alpha_mG_m(x)
  \right)
\end{align}


<br>

**5)** Para una nueva condicin de prueba $x'$, podemos predecir $y'$ por $G_{ensemble} (x')$.

<br>

**Una vez completada la secuencia (ensemble), la prediccin final viene dada por la media ponderada.**




<br>

La secuencia de **AdaBoost** se ilustra en este ejemplo.

*  Las observaciones mal clasificadas se identifican (encerradas por $\color{blue}{\text{crculos}}$, **Paso 1**)

<br>

*  Luego se les da mayor peso $w_i$ (**Paso 1'**).

<br>

*  En el siguiente paso, las observaciones que se marcaron como errores en el paso anterior se clasifican correctamente (**Paso 2**).

<br>

*  Cuando combinamos todos estos modelos o aprendices, terminamos con el aprendiz de **ensemble** que es capaz de hacer la mejor prediccin.

```python
import numpy as np
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn import metrics
from sklearn.datasets import load_breast_cancer
warnings.filterwarnings(action='ignore')            # Desactivando las advertencias del sistema.
```

###### 1.1. Leer los datos:

```python
# Cargar los datos.
data = load_breast_cancer()
```

```python
# Variables Explicativas
X = data['data']
print(data['feature_names'])
```

```python
X.shape
```

```python
# Variable de respuesta.
# Re-etiquetar de modo que 0 = 'benign' y 1 = malignant.
Y = 1 - data['target']
label = list(data['target_names'])
label.reverse()
print(label)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

```python
# rbol de Clasificacin.
DTC = DecisionTreeClassifier(max_depth=10)
DTC.fit(X_train, Y_train)
Y_pred = DTC.predict(X_test)
print( "Accuracy del rbol: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Clasificacin con KNN
KNN = KNeighborsClassifier(n_neighbors=5)
KNN.fit(X_train, Y_train)
Y_pred = KNN.predict(X_test)
print( "Accuracy de KNN: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn import metrics
from sklearn.datasets import load_breast_cancer
warnings.filterwarnings(action='ignore')                  # desactivar advertencias del sistema.
```

#### 1.1. Read in data:

```python
# Cargar los datos.
data = load_breast_cancer()
```

```python
# Variables Explicativas.
X = data['data']
variable_names = data['feature_names']
print(variable_names)
```

```python
X.shape
```

```python
# Variables de respuesta.
# Re-etiquetar de modo que 0 = 'benign' y 1 = malignant.
Y = 1 - data['target']
label = list(data['target_names'])
label.reverse()
print(label)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

###### 1.2. Random Forest:

```python
# Clasificacin por Random Forest.
RFC = RandomForestClassifier(n_estimators=100,max_depth=5,min_samples_leaf=2,random_state=123)
RFC.fit(X_train, Y_train)
Y_pred = RFC.predict(X_test)
print( "Random Forest accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Desplegar el top 10 de variables ms importantes
variable_importance = pd.Series(RFC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```

##### 2. Boosting Ensemble (Ensemble de Boosting):

###### 2.1.  AdaBoost:

```python
# Clasificacin por AdaBoost.
ABC = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=10),n_estimators=100,learning_rate=0.01,random_state=123)
ABC.fit(X_train, Y_train)
Y_pred = ABC.predict(X_test)
print( "AdaBoost accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Display the top 10 most important variables.
variable_importance = pd.Series(ABC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```

###### 2.2.  Gradient Boosting (Impulso de Gradiente):

```python
# Clasificacin por Gradient Boosting.
GBC = GradientBoostingClassifier(n_estimators=100,learning_rate=0.01,min_samples_leaf=2,max_leaf_nodes=30,random_state=123)
GBC.fit(X_train, Y_train)
Y_pred = GBC.predict(X_test)
print( "Gradient Boosting accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Mostrar el top 10 de variables ms importantes.
variable_importance = pd.Series(GBC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```

###### 2.3. XGBoost:

```python
# Clasificacin por XGBoost.
XGBC = XGBClassifier(n_estimator = 500, learning_Rate = 0.1, max_depth = 4, random_state=123)
XGBC.fit(X_train, Y_train)
Y_pred = XGBC.predict(X_test)
print( "XGBoost accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Mostrar el top 10 de variables ms importantes.
variable_importance = pd.Series(XGBC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```
