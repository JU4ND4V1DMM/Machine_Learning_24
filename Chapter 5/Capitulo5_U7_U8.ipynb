## **Unidad 7. Algoritmo SVM**

### **| 7.1 Algoritmo SVM**

$\color{blue}{\text{|}}$ **A cerca de SVM (Máquina de Soporte Vectorial o Supported Vector Machine)**


$\color{blue}{\to}$ Mejora de la precisión de la clasificación al maximizar el margen

$\color{blue}{\to}$ Límite de clasificación no lineal efectivo por la transformación "kernel"

<br>

SVM introdujo muchos conceptos nuevos.

Se trata de un algoritmo muy eficaz que puede funcionar bien incluso cuando se requiere un límite de clasificación no lineal.




<br>

$\color{blue}{\to}$ El SVM es un modelo que clasifica los datos encontrando la línea (o hiperplano) donde la distancia (margen) entre los datos en diferentes categorías se vuelve máxima.

<br>

$\color{blue}{\to}$ Como se muestra en la figura siguiente, el modelo de SVM encuentra el hiperplano que divide los datos en dos categorías diferentes con el margen máximo para clasificar los datos.  

<br>

$\color{blue}{\to}$ Habría muchas líneas o planos que dividen los datos en dos categorías diferentes. Sin embargo, los puntos que superan el límite de clasificación pueden aparecer en datos de evaluación o datos futuros desconocidos a medida que los puntos cercanos al límite cambian ligeramente.


<br>

 SVM es un algoritmo potente. Sin embargo, se necesita más tiempo y esfuerzo para entrenarlo adecuadamente.


<br>

$\color{blue}{\to}$ Para minimizar esta posibilidad, se debe encontrar una línea (o hiperplano) que haga el máximo margen entre las diferentes categorías de datos. En otras palabras, el objetivo es encontrar un hiperplano que pueda inducir la máxima diferenciación para la mejor generalización posible para clasificar y predecir los datos futuros, no los datos de entrenamiento actuales.

<br>

$\color{blue}{\to}$ El punto de la categoría más cercano a la línea límite se denomina vector de soporte, y cada clasificación debe tener al menos un vector de soporte.

<br>


<br>

$\color{blue}{\to}$ Sin embargo, no siempre es posible clasificar todos los datos de forma lineal. A veces, la clasificación de datos debe realizarse en una curva o en un plano de clasificación no lineal más complejo.

<br>

$\color{blue}{\to}$ Si es así, utiliza el método de truco Kernel para asignar los datos especificados a una dimensión superior y busque un hiperplano que pueda clasificar los datos en la dimensión convertida.

<br>

$\color{blue}{\to}$ En lugar de convertir los datos a una dimensión superior, el método de truco Kernel utiliza la función Kernel que devuelve un valor similar al que se realiza cuando se realizan cálculos internos entre vectores en la dimensión superior. Esto puede dar lugar a una clasificación no lineal sin mover los datos a una dimensión superior.

<br>

$\color{blue}{\to}$ Las principales funciones del kernel incluyen núcleos polinómicos, gaussianos y sigmoides.




<br>

$\color{blue}{\text{|}}$ **Pros**

$\color{blue}{\to}$ No es muy sensible a los valores atípicos.

$\color{blue}{\to}$ El rendimiento es bueno.

<br>

$\color{blue}{\text{|}}$ **Contras**

$\color{blue}{\to}$ El entrenamiento es relativamente lento. Funciona mal para datos de gran tamaño.

$\color{blue}{\to}$ El kernel y el conjunto de hiperparámetros deben optimizarse cuidadosamente.

$\color{blue}{\to}$ No se puede obtener mucha información.

$\color{blue}{\text{|}}$ **Hiperplano**

$\color{blue}{\to}$ Para un espacio configuracional k dimensional, el hiperplano tiene la dimensión k−1.

$\color{blue}{\to}$ Aquí, expresamos matemáticamente el "hiperplano" que divide en dos el espacio configuracional.

<br>

**Ejemplo.**

Para un espacio bidimensional, un hiperplano es una línea bisectrizable que se puede parametrizar como:


\begin{align}
\large
\beta_0+\beta_1X_1+\beta_2X_2=0
\end{align}

<br>

El espacio bidimensional se subdivide en dos:

\begin{align}
\large
\beta_0+\beta_1X_1+\beta_2X_2>0 \\
\large
\beta_0+\beta_1X_1+\beta_2X_2<0
\end{align}

<br>

Una observación pertenecería a cualquiera de ellos $\Rightarrow$ clasificación binaria!


<br>

$\color{blue}{\text{|}}$ **Optimización de márgenes (para el binario $y$)**

SVM consiste en optimizar los parámetros del hiperplano de manera que se maximice el margen.
Al ajustar el hiperparámetro $C$, podemos controlar la "flexibilidad".

Aunque la explicación se limita a la respuesta binaria ($y$), SVM se puede generalizar a la predicción multiclase utilizando el esquema "uno contra todos".

<br>

$\color{blue}{\to}$ El margen de límite se maximiza.

$\color{blue}{\to}$ Si no es posible establecer un margen claro, se permite un error dentro de un límite.

$\color{blue}{\to}$ El objetivo es maximizar $M$ optimizando los parámetros $\beta_0,\beta_1,\cdots,\beta_k$ sujetos a las restricciones:

<br>

*  **Restricción 1:** $\sum_{j=1}^k \beta_j^2 =1$

<br>

*  **Restricción 2:** $y_i (\beta_0+\beta_1 x_{i1}+\beta_2 x_{i2}+...+\beta_k x_{ik} )≥M\color{red}{(1-\varepsilon_i)}, i=1,2,\cdots,n \\
\\
y_i \text{ es -1 o +1.}$



<br>

*  **Restricción 3:** $\color{red}{\varepsilon_i \ge 0}$ y $\color{red}{\sum_{i=1}^k\varepsilon_i \le C}$ $\color{green}{\text{ } \Leftarrow}$ $C$ *es un hiperparámetro relacionado con los errores de clasificación*


<br>

<br>

$\color{blue}{\to}$ Mapeo a una dimensión superior usando las funciones "kernel".

$\color{blue}{\Rightarrow}$ En 1D, es obviamente imposible encontrar un límite lineal para la clasificación correcta.

<br>

$\color{blue}{\Rightarrow}$ En 2D, es imposible encontrar un límite lineal para la clasificación correcta.

<br>

$\color{blue}{\to}$ Mapeo efectivo a una dimensión superior al dar el producto interno de dos vectores $x_1$ y $x_2$.

<br>

La función kernel da el producto interno en dimensiones más altas.

Hay varios tipos diferentes de funciones del kernel.



<br>

*  Lineal: $K(x_1,x_2 )=x_1^t x_2$

*  Polinomio: $K(x_1,x_2 )=(x_1^t x_2+c)^d$, donde $c>0$

*  Sigmoide: $K(x_1,x_2 )=tanh(a(x_1^t x_2 )+b)$, donde $a, b>0$

*  Base de la función radial (rbf): $K(x_1,x_2 )=exp(-\gamma‖x_1-x_2 ‖^2 )$, donde $\gamma>0$

##### 1. Clasificación con SVM:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import recall_score
from sklearn.datasets import load_iris
warnings.filterwarnings(action='ignore')                  # Desactivar las advertencias del sistema.
%matplotlib inline
```

###### 1.1. Leer los datos:

```python
# Cargar los datos.
data = load_iris()
```

```python
# Variables explicativas.
X = data['data']
columns = list(data['feature_names'])
print(columns)
```

```python
# Variable de respuesta.
Y = data['target']
labels = list(data['target_names'])
print(labels)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

###### 1.2. Optimización de hiperparámetros SVM (RBF kernel):

C     : Parámetro de penalidad.
<br>

gamma : parámetro kernel ($\gamma$).

```python
C_grid = 0.02*np.arange(20,50)
gamma_grid = 0.02*np.arange(1,50)
parameters = {'C': C_grid, 'gamma' : gamma_grid}
gridCV = GridSearchCV(SVC(kernel='rbf'), parameters, cv=10, n_jobs=-1)          # "n_jobs = -1" significa "use todos los CPU cores".
gridCV.fit(X_train, Y_train)
best_C = gridCV.best_params_['C']
best_gamma = gridCV.best_params_['gamma']
```

```python
print("SVM mejor C : " + str(best_C))
print("SVM mejor gamma : " + str(best_gamma))
```

```python
SVM_best = SVC(kernel='rbf', C=best_C, gamma=best_gamma)
SVM_best.fit(X_train, Y_train);
Y_pred = SVM_best.predict(X_test)
print( "SVM mejor accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
print( "SVM mejor recall : " + str(np.round(metrics.recall_score(Y_test,Y_pred, average='weighted'),3)))
```

###### 1.3. Optimización de hiperparámetro SVM (kernel polinomial):

```python
C_grid = 0.0001*np.arange(1,100)
gamma_grid = 0.01*np.arange(1,100)
parameters = {'C': C_grid, 'gamma' : gamma_grid}
gridCV = GridSearchCV(SVC(kernel='poly'), parameters, cv=10, n_jobs=-1)         # "n_jobs = -1" significa "use todos los CPU cores".
gridCV.fit(X_train, Y_train)
best_C = gridCV.best_params_['C']
best_gamma = gridCV.best_params_['gamma']
```

```python
print("SVM mejor C : " + str(best_C))
print("SVM mejor gamma : " + str(best_gamma))
```

```python
SVM_best = SVC(kernel='poly', C=best_C,gamma=best_gamma)
SVM_best.fit(X_train, Y_train);
Y_pred = SVM_best.predict(X_test)
print( "SVM mejor accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

**Algoritmos Ensemble**

<br>

$\color{blue}{\text{|}}$ **A cerca de Algoritmos Ensemble**

Hasta ahora, hemos cubierto varios algoritmos de aprendizaje automático.

Ahora, consideramos los llamados algoritmos de conjunto que se basan en otros modelos relativamente débiles.

Los principales tipos son los que se explican aquí.

<br>


$\color{blue}{\to}$ Modelo predictivo fuerte basado en los modelos más débiles.

<br>

$\color{blue}{\to}$ $\color{blue}{\text{Tipo de votación:}}$

*  Un conjunto de modelos básicos que "votan"

*  Un conjunto de diferentes tipos de modelos. **Ejemplo.** Combina Árbol de Decisión, KNN, SVM, etc.

<br>

$\color{blue}{\to}$ $\color{blue}{\text{Tipo de embolsado:}}$

*  Una colección de modelos débiles e independientes que "votan"

*  Un conjunto del mismo tipo de modelos débiles. **Ejemplo.** Random Forest

<br>

$\color{blue}{\to}$ $\color{blue}{\text{Tipo de impulso:}}$

*  Una serie de modelos débiles que aprenden y predicen adaptativamente. **Ejemplo.** AdaBoost, GBM, XGBoost, etc.

*  La serie se hace crecer añadiendo nuevos modelos multiplicados por los "boosting weights" o "pesos de aumento".

<br>

$\color{blue}{\to}$ Tomar una decisión más adecuada combinando adecuadamente múltiples opiniones obtenidas por diferentes expertos

<br>

$\color{blue}{\to}$ Votación por mayoría: análisis utilizando muchos modelos de clasificación diferentes con un conjunto de entrenamiento idéntico.

<br>

$\color{blue}{\to}$ Embolsado (conjunto de entrenamiento diferente)
*  Bootstrap + agregación
*  Bootstrap + agregación
*  Embolsado (muestreo aleatorio con reemplazo (arranque aleatorio)). **Ejemplo.**Random Forest (Bosque Aleatorio).

<br>

$\color{blue}{\to}$ Boosting (Impulso)

*  Conserve el 50% de los datos clasificados incorrectamente o utilice el peso para la selección de muestras.

<br>

Es bastante fácil implementar un Ensemble de votación con la librería scikit-learn.






<br>

$\color{blue}{\text{|}}$ **Boostrap**

$\color{blue}{\to}$ Estimación de estadísticas desconocidas

$\color{blue}{\to}$ Método de estimación fácil y eficaz con la distribución desconocida de la muestra de parámetros del modelo

$\color{blue}{\to}$ Proceso de recálculo de los estadísticos y el modelo para cada muestra a través de muestreos adicionales con reemplazo de la muestra actual
*  No se requiere ninguna suposición de que los parámetros o los estadísticos de la muestra deben estar en una distribución normal.

$\color{blue}{\to}$ bootstrap sample - Agregación de los datos observados (muestreo con reemplazo obtenido a partir del valor de registro y la variable dependiente)

<br>

$\color{blue}{\text{Contenidos a aprender en el futuro}}$

*  Remuestreo: implica un submuestreo de permutación sin reemplazo.
*  Agregación de bootstrap: obtención de un resultado a partir de la agregación de valores predichos obtenidos de diferentes muestras de bootstrap.



<br>

$\color{blue}{\text{|}}$ **¿Qué es aprendizaje Ensemble?**

<br>

$\color{blue}{\to}$ El término *Ensemble* puede definirse de la siguiente manera:
*  En mecánica estadística, un *ensemble* de un sistema se refiere a la colección de sistemas equivalentes.

<br>

$\color{blue}{\to}$ Es decir, es un conjunto de grupos similares.

<br>

$\color{blue}{\to}$ En lugar de esperar resultados de rendimiento de un solo modelo, el aprendizaje  ensemble obtiene un mejor resultado utilizando la inteligencia colectiva de diferentes modelos, como promediar muchos modelos individuales diferentes o tomar una decisión basada en el voto de la mayoría.

<br>

$\color{blue}{\to}$ Hay muchos métodos de ensamble diferentes que utilizan la inteligencia colectiva.

*  Votación: obtención de resultados a través de la votación
*  Embolsado: agregación de Bootstrap (creación duplicada de varias muestras)
*  Boosting - Ponderación complementando errores anteriores
*  Apilamiento: un metamodelo basado en diferentes modelos

<br>

$\color{blue}{\to}$ Puede haber otros métodos diferentes más, ya que el aprendizaje por ensemble aplica una determinada técnica/metodología. Sin embargo, los cuatro métodos enumerados anteriormente son las técnicas de ensemble más representativas de la librería sklearn.


<br>

$\color{blue}{\text{|}}$ **Ensemble de votación**

$\color{blue}{\to}$ Se puede aplicar a la clasificación y la regresión.

<br>

$\color{blue}{\to}$ Los modelos que forman un ensemble deben ser de diferentes tipos para un buen desempeño.

<br>

$\color{blue}{\to}$ Dos métodos de votación para el clasificador:
*  **Duro**: la etiqueta de clase predicha viene dada por la votación de la regla de la mayoría.

*  **Suave**: la etiqueta de clase predicha viene dada por el $\color{blue}{\text{argmax}}$ de la suma de las probabilidades predichas.

<br>

Podemos hacer un ensemble de votación en el que los modelos básicos deben ser de diferentes tipos para un buen desempeño.

El *argmax* significa la posición en el valor más grande.

$\color{blue}{\text{|}}$ **Votación**

$\color{blue}{\to}$ Es bastante fácil implementar un ensemble de votación con la librería scikit-learn.

<br>

$\color{blue}{\to}$ Como la palabra misma, votación consiste en tomar una decisión a través de los votos. La votación es similar al (bagging) embolsado, ya que utiliza un método de votación, pero se diferencian entre sí de la siguiente manera:

<br>

*  **Votación**: Combina diferentes modelos de algoritmos.
*  **Embolsado (bagging)**: Utiliza diferentes combinaciones de muestras dentro del mismo algoritmo.

<br>

$\color{blue}{\to}$ La votación selecciona los resultados finales haciendo que la votación final sobre los resultados sea deducida por diferentes algoritmos.

<br>

$\color{blue}{\to}$ La votación se clasifica en voto duro y voto blando.

<br>

*  **Votación dura**: Decide el valor final del resultado a través de la votación.

*  **Votación suave**: Extrae el valor final sumando todos los valores de probabilidad de obtener el resultado final y luego calculando cada probabilidad del resultado final.

<br>

$\color{blue}{\text{|}}$ **Partes del sistema de votación**

SIC2023_C5_U8_384.png


$\color{blue}{\to}$ **Votación dura**

*  Tomando como ejemplo la clasificación. Supongamos que los valores predictivos arrojados por los **modelos individuales** para la clasificación son 1,0,0,1,1, ya que la $\color{blue}{\text{clase 1}}$ tiene tres votos y la $\color{red}{\text{clase 0}}$ tiene 2 votos. En ese caso, 1 se convierte en el valor predictivo final en el método de votación dura.

<br>

$\color{blue}{\to}$ **Votación blanda**

*  El método de votación blanda calcula el valor promedio de cada probabilidad y luego determina la que tiene la probabilidad más alta.

*  Supongamos que los **modelos individuales** arrojan la probabilidad de obtener la $\color{red}{\text{clase 0}}$ es $\color{red}{(0.4, 0.9, 0.9, 0.4, 0.4)}$

  y la probabilidad de obtener la $\color{blue}{\text{clase 1}}$ es $\color{blue}{(0.6, 0.1, 0.1, 0.6, 0.6)}$.

<br>

*  En base a la **votación blanda** La probabilidad final de obtener la $\color{red}{\text{clase 0}}$ es
$\large\frac{(0,4 + 0,9 + 0,9 + 0,4 + 0,4)}{5} = 0,6$;

  La probabilidad final de obtener la $\color{blue}{\text{clase 1}}$ es
  $\large\frac{$(0,6 + 0,1 + 0,1 + 0,6 + 0,6)}{5} = 0,4$.
  
  Por lo tanto, el valor final seleccionado es diferente del resultado de la votación dura anterior.

<br>

$\color{blue}{\to}$ En general, el uso del método de votación blanda se considera más razonable que el método de votación dura en las competiciones, ya que el método de votación blanda proporciona un resultado de rendimiento real mucho mejor.

**Ensemble de embolsado (Bagging Ensemble): Bosque Aleatorio**

$\color{blue}{\text{|}}$ **A cerca del Bosque Aleatorio**

Bosque Aleatorio o Random Forest es un algoritmo de conjunto basado en los aprendices de árbol ("aprendices débiles").

Los pros y los contras se enumeran aquí

<br>

$\color{blue}{\to}$ Un algoritmo Ensemble basado en árboles

$\color{blue}{\to}$ Puede ser aplicado para clasificación y regresión

<br>

$\color{blue}{\text{|}}$ **Pros**

$\color{blue}{\to}$ Poderosa

$\color{blue}{\to}$ Pocas asumpciones

$\color{blue}{\to}$ Poca o no preocupación sobre el problema de sobreajuste (overfitting)

<br>

$\color{blue}{\text{|}}$ **Contras**

$\color{blue}{\to}$ El entrenamiento toma tiempo

<br>

$\color{blue}{\text{|}}$ **Algoritmo de Bosque Aleatorio**

Como se mencionó anteriormente, Random Forest es un algoritmo de conjunto basado en los aprendices de Arboles.

Los valores pronosticados vienen dados por el voto mayoritario.


SIC2023_C5_U8_390.png

$\color{blue}{\to}$ Los árboles elegidos al azar forman un bosque

$\color{blue}{\to}$ La predicción se decide por el voto mayoritario


<br>

Cada árbol que compone un conjunto de bosque aleatorio se entrena utilizando un conjunto de observaciones y variables seleccionadas aleatoriamente.


<br>

Pasos del sistema:


**1)** Hacer árboles con variables y observaciones seleccionadas al azar.

**2)** Mantener solo aquellos con la impureza (o entropía) de Gini más baja.

**3)** Repita desde el paso **1)** un número determinado de veces.

### **|8.3. Impulso (Boosting)**

<br>

$\color{blue}{\text{|}}$ **Conjunto de impulso: AdaBoost**

**Acerca de AdaBoost**

$\color{blue}{\to}$ Una secuencia de modelos débiles, como los árboles

<br>

$\color{blue}{\to}$ Un conjunto ponderado de modelos débiles

<br>

$\color{blue}{\to}$ Un conjunto de ponderaciones que aumentan la importancia de los modelos con mejor rendimiento

<br>

$\color{blue}{\to}$ Un conjunto de ponderaciones que aumentan la importancia de las observaciones erróneamente clasificadas

<br>

$\color{blue}{\to}$ Pros y contras similares a los del Random Forest

<br>

$\color{blue}{\to}$ AdaBoost es un conjunto de modelos débiles (generalmente árboles).

<br>

$\color{blue}{\to}$ A diferencia de Random Forest, donde los modelos básicos se entrenan a la vez en paralelo, AdaBoost tiene una secuencia de modelos básicos que se entrenan uno tras otro.

<br>


$\color{blue}{\text{|}}$ **Algoritmo de Clasificación AdaBoost**


$\color{blue}{\to}$ Supongamos:

*  $n$ observaciones para el paso de entrenamiento: $x_i$ y $y_i$.

*  También suponemos que $y_i \in \{−1,+1\}$.  ($y$ binario)

*  La predicción se basa en el uso de la función $sign()$ que devuelve +1 o −1.




<br>

$\color{blue}{\to}$ Haremos una serie de modelos débiles $G_m (x)$ con $m= 1,\cdots,M$.

<br>

$\color{blue}{\to}$ El clasificador *ensemble* se compone de una combinación lineal de estos modelos débiles:


\begin{align}
\large
G_{ensemble}(x)=sign
\left(  \sum_{m=1}^M\alpha_mG_m(x)
  \right)
\end{align}

<br>

donde $\alpha_m$ son los "pesos de aumento" o "boosting weights" que deben calcularse.


$\color{blue}{\to}$ Le asignamos un mayor peso a un modelo de mejor rendimiento.

<br>

El ensemble en AdaBoost es, en realidad, una secuencia.



<br>

$\color{blue}{\to}$ Cada paso se compone de entrenamiento y predicción.

<br>

$\color{blue}{\to}$ Se entrena a un modelo básico y, a continuación, se estiman los coeficientes de error de la predicción.


**1)** Para el primer paso ($m=1$), se asigna el mismo peso a las observaciones:

\begin{align}
w_i^{(1)}\frac{-1}{n}
\end{align}

<br>

**2)** Para la secuencia de boos $m = 1,\cdots, M$:

  **a)** Entrena al modelo $G_m (x)$ utilizando observaciones ponderadas por $w_i^{(m)}$.  

  **b)** Calcula la tasa de error:

\begin{align}
\epsilon_m=\frac{\sum_{i=1}^nw_i^{(m)}\color{red}{I(y_i \ne G_m(x_i))}}{\sum_{i=1}^nw_i^{(m)}}
\end{align}

<br>

$\Rightarrow \color{red}{{I(y_i \ne G_m(x_i))}}$ resulta 1 para una predicción incorrecta, sino resulta 0.

$\Rightarrow 0 \le \epsilon_m \le 1$



<br>

**3)** Para la secuencia de refuerzo $m = 1,\cdots, M$:

**c)** Calcule el peso de *boosting*:

\begin{align}
\alpha_m=\frac{1}{2}log\left(\frac{1-\epsilon_m}{\epsilon_m}\right)
\end{align}

<br>

$\Rightarrow$ Como $\epsilon_m\to0$, $\alpha_m$ es un número positivo grande. ¡Al modelo se le da más importancia!

$\Rightarrow$  Como $\epsilon_m\cong0.5$, $\alpha_m\cong0$.

$\Rightarrow$  Como $\epsilon_m\to1$, $\alpha_m$ es un número negativo grande.


<br>

d) Para el siguiente paso, los pesos de la observación predicha $\color{red}{\text{erróneamente}}$ se reescalan mediante un factor $e_m^a$.  

Esto se puede expresar de forma compacta como:

\begin{align}
w_i^{(m+1)}=w_i^{(m)}\times e^{\alpha \cdot m \cdot I(y_i \ne G_m(x_i))}\text{, donde }i=1,\cdots,n
\end{align}

<br>

$\Rightarrow$ En el siguiente paso de la secuencia, las observaciones predichas $\color{red}{\text{erróneamente}}$ reciben un mayor peso.

<br>



**4)** El clasificador ensemble se compone de una combinación lineal de modelos débiles:

\begin{align}
\large
G_{ensemble}(x)=sign
\left(  \sum_{m=1}^M\alpha_mG_m(x)
  \right)
\end{align}


<br>

**5)** Para una nueva condición de prueba $x'$, podemos predecir $y'$ por $G_{ensemble} (x')$.

<br>

**Una vez completada la secuencia (ensemble), la predicción final viene dada por la media ponderada.**




<br>

La secuencia de **AdaBoost** se ilustra en este ejemplo.

*  Las observaciones mal clasificadas se identifican (encerradas por $\color{blue}{\text{círculos}}$, **Paso 1**)

<br>

*  Luego se les da mayor peso $w_i$ (**Paso 1'**).

<br>

*  En el siguiente paso, las observaciones que se marcaron como errores en el paso anterior se clasifican correctamente (**Paso 2**).

<br>

*  Cuando combinamos todos estos modelos o aprendices, terminamos con el aprendiz de **ensemble** que es capaz de hacer la mejor predicción.

<br>

$\color{blue}{\to}$ Loss = Y_predict - Y_actual

```python
import numpy as np
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn import metrics
from sklearn.datasets import load_breast_cancer
warnings.filterwarnings(action='ignore')            # Desactivando las advertencias del sistema.
```

###### 1.1. Leer los datos:

```python
# Cargar los datos.
data = load_breast_cancer()
```

```python
# Variables Explicativas
X = data['data']
print(data['feature_names'])
```

```python
X.shape
```

```python
# Variable de respuesta.
# Re-etiquetar de modo que 0 = 'benign' y 1 = malignant.
Y = 1 - data['target']
label = list(data['target_names'])
label.reverse()
print(label)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

###### 1.2. Prediciendo con estimador individual:

```python
# Árbol de Clasificación.
DTC = DecisionTreeClassifier(max_depth=10)
DTC.fit(X_train, Y_train)
Y_pred = DTC.predict(X_test)
print( "Accuracy del Árbol: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Clasificación con KNN
KNN = KNeighborsClassifier(n_neighbors=5)
KNN.fit(X_train, Y_train)
Y_pred = KNN.predict(X_test)
print( "Accuracy de KNN: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Classification with Logistic Regression.
LR = LogisticRegression()
LR.fit(X_train, Y_train)
Y_pred = LR.predict(X_test)
print( "Accuracy de Regresión Logistica: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

###### 1.3. Prediciendo con un ensemble de votación:

```python
VC = VotingClassifier(estimators=[('Tree',DTC),('knn',KNN),('Logistic',LR)],voting='hard')      # voting = 'hard'.
VC.fit(X_train, Y_train)
Y_pred = VC.predict(X_test)
print( "Accuracy de Clasificador por Votación con Voting=Hard: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
VC = VotingClassifier(estimators=[('Tree',DTC),("KNeighbor",KNN),("LogisticReg",LR)],voting='soft')     # voting = 'soft'.
VC.fit(X_train, Y_train)
Y_pred = VC.predict(X_test)

print( "Accuracy de Clasificador por Votación con Voting=Soft: " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn import metrics
from sklearn.datasets import load_breast_cancer
warnings.filterwarnings(action='ignore')                  # desactivar advertencias del sistema.
```

#### 1.1. Read in data:

```python
# Cargar los datos.
data = load_breast_cancer()
```

```python
# Variables Explicativas.
X = data['data']
variable_names = data['feature_names']
print(variable_names)
```

```python
X.shape
```

```python
# Variables de respuesta.
# Re-etiquetar de modo que 0 = 'benign' y 1 = malignant.
Y = 1 - data['target']
label = list(data['target_names'])
label.reverse()
print(label)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

###### 1.2. Random Forest:

```python
# Clasificación por Random Forest.
RFC = RandomForestClassifier(n_estimators=100,max_depth=5,min_samples_leaf=2,random_state=123)
RFC.fit(X_train, Y_train)
Y_pred = RFC.predict(X_test)
print( "Random Forest accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Desplegar el top 10 de variables más importantes
variable_importance = pd.Series(RFC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```

##### 2. Boosting Ensemble (Ensemble de Boosting):

###### 2.1.  AdaBoost:

```python
# Clasificación por AdaBoost.
ABC = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=10),n_estimators=100,learning_rate=0.01,random_state=123)
ABC.fit(X_train, Y_train)
Y_pred = ABC.predict(X_test)
print( "AdaBoost accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Display the top 10 most important variables.
variable_importance = pd.Series(ABC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```

###### 2.2.  Gradient Boosting (Impulso de Gradiente):

```python
# Clasificación por Gradient Boosting.
GBC = GradientBoostingClassifier(n_estimators=100,learning_rate=0.01,min_samples_leaf=2,max_leaf_nodes=30,random_state=123)
GBC.fit(X_train, Y_train)
Y_pred = GBC.predict(X_test)
print( "Gradient Boosting accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Mostrar el top 10 de variables más importantes.
variable_importance = pd.Series(GBC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```

###### 2.3. XGBoost:

```python
# Clasificación por XGBoost.
XGBC = XGBClassifier(n_estimator = 500, learning_Rate = 0.1, max_depth = 4, random_state=123)
XGBC.fit(X_train, Y_train)
Y_pred = XGBC.predict(X_test)
print( "XGBoost accuracy : " + str(np.round(metrics.accuracy_score(Y_test,Y_pred),3)))
```

```python
# Mostrar el top 10 de variables más importantes.
variable_importance = pd.Series(XGBC.feature_importances_, index = variable_names)
top_variables_10 = variable_importance.sort_values(ascending=False)[:10]
sns.barplot(x=top_variables_10.values, y = top_variables_10.index, ci=None)
plt.show()
```
