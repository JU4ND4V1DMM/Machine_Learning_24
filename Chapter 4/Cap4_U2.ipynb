## **Densidad de probabilidad continua**

* Variable aleatoria continua y `densidad` de probabilidad.

* Nmero infinito `no numerable` ($\aleph_1  $) de valores posibles.

* La probabilidad en un valor especfico es cero: $P(X = x_0) = 0$

* Probabilidad es distinta de cero solo para intervalos: $P(x_1 \leq X \leq x_2) = \int_{x_1}^{x_2} f(x) \ dx\ $

* Probabilidad acumulada: $ CDF(x) = P(-\infty < X \leq x) = \int_{-\infty}^{x} f(y) \ dy\ = F(x)$

* Podemos calcular $P(x_1 \leq X \leq x_2) = CDF(x_2) - CDF(x_1) = F(x_2) - F(x_1)$

</br>

2.2. Probabilidad conjunta

## Covarianza

* Se trata de dos variables X y Y con una probabilidad conjunta implcita.
* Se puede calcular como valor esperado: $Cov(X,Y) = E[(X-\mu_x)(Y-\mu_y)]$.
* Una expresin bastante ms simple: $Cov(X,Y) = E[X Y]-E[X]E[Y]$.
* Cuando X=Y, tenemos $Cov(X,X) = Var(X)$.

<FONT COLOR="gray">
Nota: La probabilidad conjunta (o densidad de probabilidad) se requiere cuando se calcula la covarianza o correlacin entre dos variables aleatorias.

- La covarianza puede ser positiva, negativa o cero. Una covarianza positiva indica que las variables tienden a moverse en la misma direccin, mientras que una covarianza negativa indica que las variables tienden a moverse en direcciones opuestas. Una covarianza cero indica que las variables no estn relacionadas linealmente.
- La covarianza se puede utilizar para estimar la pendiente de la recta de regresin entre dos variables.
- La covarianza tambin se puede utilizar para determinar si dos variables son independientes. Si la covarianza es cero, entonces las variables son independientes.
</FONT>

2.2. Probabilidad conjunta

## Correlacin

La correlacin y la covarianza estn relacionadas:

$Cor(X,Y)=\frac{Cov(X,Y)}{\sigma_x\sigma_y}$

Valor en el intervalo [1,1]
La correlacin es una medida de la relacin lineal entre X y Y.

a) $Cor(X,Y)$: una relacin lineal <FONT COLOR="red">positiva</FONT> entre X y Y

b) $Cor(X,Y)$: una relacin lineal <FONT COLOR="blue">negativa</FONT>  entre X y Y

c) $Cor(X,Y)$: <FONT COLOR="green">no</FONT> hay relacin lineal entre X y Y

<FONT COLOR="gray">
Nota: La correlacin es una medida de la relacin lineal entre dos variables aleatorias.

## Dependencia y correlacin

Si <FONT COLOR="red">X</FONT> y <FONT COLOR="blue">Y</FONT> son independientes: $P($<FONT COLOR="red">X</FONT>, <FONT COLOR="blue">Y</FONT>) = $P($<FONT COLOR="red">X</FONT> )$P($<FONT COLOR="blue">Y</FONT>) o $f($<FONT COLOR="red">X</FONT> , <FONT COLOR="blue">Y</FONT>) = $f($<FONT COLOR="red">X</FONT> )$f($<FONT COLOR="blue">Y</FONT>).

Entonces, $Cov($<FONT COLOR="red">X</FONT>,<FONT COLOR="blue">Y</FONT>) = E[<FONT COLOR="red">X</FONT> <FONT COLOR="blue">Y</FONT>]- E[<FONT COLOR="red">X</FONT>] E[<FONT COLOR="blue">Y</FONT>] = E[<FONT COLOR="red">X</FONT>] E[<FONT COLOR="blue">Y</FONT>]- E[<FONT COLOR="red">X</FONT>] E[<FONT COLOR="blue">Y</FONT>]=0

Entonces, independencia **siempre** significa falta de correlacin.
