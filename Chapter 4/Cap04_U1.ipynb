### **Definición estadística (empírica o frecuentista) de probabilidad**

*   El número total de observaciones de un experimento aleatorio es $N$.
*   El número total de observaciones que corresponden al evento $A$ es $N_A$.
*   Idealmente, la probabilidad está dada por el siguiente límite:

\begin{align}
\large P(A)= \lim_{N \to \infty} \frac{N_A}{N}
\end{align}

*   Sin embargo, el siguiente es más realista:

\begin{align}
\large P(A)=\frac{N_A}{N}
\end{align}

**Ejemplo.** Según los datos, de 1,000 recién nacidos, 985 viven después del primer año.
Cuál es la probabilidad de supervivencia en un año?

\begin{align}
P=\frac{980}{1000}=0.98
\end{align}

**Ejemplo.** Un dado fue lanzado 10,000 veces de las cuales 1,650 veces el resultado "uno" fue obtenido.
¿Cuál es la probabilidad del resultado "uno"?

\begin{align}
\large P=\frac{1650}{10000}=0.165\approx\frac{1}{6}
\end{align}

Aquí aplicamos la definición empírica de probabilidad que vimos anteriormente.

**Probabilidad Condicional**

**Ejemplo.**

En una bolsa hay dos bolas blancas y tres bolas rojas.
Suponga que dos bolas son tomadas una tras otra.
Calcule la probabilidad de que las dos bolas tomadas sean blancas.

<br>

a) Suponga que la primera bola **no es devuelta** a la bolsa.

Denotemos con $A=$ evento que la primera bola es blanca
y $B=$ evento que la segunda bola es blanca.

\begin{align}
P(A \cap B)=P(B|A)P(A)=\frac{1}{4}\times\frac{2}{5}=\frac{1}{10}
\end{align}


Cuando la primera bola **no es puesta** de vuelta en la bolsa, es relevante de que color era esta bola cuando calculamos las probabilidades de la segunda bola.

Esta relación de causa es expresada usando la probabilidad condicional.


**Ejemplo.**

En una bolsa hay dos bolas blancas y tres bolas rojas.
Suponga que dos bolas son sacadas una tras otra.
Calcule la probabilidad de que dos bolas sean blancas.

b) Suponga que la primera bola blanca es **colocada de vuelta** en la bolsa, antes de tomar la segunda bola.
Esto es como un "reinicio". Los eventos $A$ y $B$ son **independientes entre sí**.


\begin{align}
P(A \cap B)=P(B|A)P(A)=P(B)P(A)=\frac{2}{5}\times\frac{2}{5}=\frac{4}{25}
\end{align}

Cuando la primera bola es colocada **de vuelta** en la bolsa, la conexión causal ser pierde.

Entonces, la probabilidad condicional no se requiere más.

Tomar la primera bola y tomar la segunda bola se vuelven eventos completamente independientes.

Para los eventos dados $A$ y $B$ puedes pensar en los siguientes casos:

<br>

1) Si los eventos $A$ y $B$ son **dependientes** entre sí, entonces:

\begin{align}
P(A \cap B) = P(A|B)P(B) \\
= P(B|A)P(A) \\
y \\
\end{align}


\begin{align}
P(A \cap B) \ne P(A)P(B)
\end{align}

<br>

2) Si los eventos $A$ y $B$ son **independientes** entre sí, entonces:

\begin{align}
P(A \cap B) = P(B)P(A)
\end{align}

<br>

3) Si los eventos $A$ y $B$ son **mutuamente excluyentes**, entonces:

\begin{align}
P(A \cap B)=0
\end{align}

<br>

Es importante no confundir eventos **mutuamente excluyentes** con eventos **independientes**. Son conceptos completamente diferentes.

### **Teorema de Bayes**

Recordemos la siguiente relación:

\begin{align}
P(A \cap B) = P(A|B)P(B) \\
=P(B|A)P(A)
\end{align}

<br>

Usando la relación anterior, podemos mostrar el conocido **Teorema de Bayes**.

\begin{align}
P(A|B)P(B)=P(B|A)P(A)
\end{align}

<br>

Este teorema parece simple, pero hay una gran cantidad de consequencias a partir de este teorema.

Incluso existe una rama de la estadística llamada "Estadística Bayesiana".

Podemos transformar el Teorema de Bayes en la siguiente forma:

\begin{align}
P(A|B)P(B)=P(B|A)P(A)
\end{align}
\begin{align}
\bigg\downarrow
\end{align}

\begin{align}
P(A|B)=\frac{P(B|A)P(A)}{P(B)} \\
\frac{P(B|A)P(A)}{P(B|A)P(A)+P(B|A^c)P(A^c)}
\end{align}

Hemos transformado el Teorema de Bayes en un formato más "útil".

Preste especial atención al denominador.

Por qué no calcular solo el $P(B)$ en el denominador sin adentrarnos en todo este problema?

Porque no se puede. Ya veremos por qué.

Podemos transformar el teorema de Bayes en la siguiente forma:

\begin{align}
P(A|B)=\frac{P(B|A)P(A)}{P(B|A)P(A)+P(B|A^c)P(A^c)}
\end{align}

<br>

La $P(B)$ no puede ser calculada de otra forma porque el evento $B$ depende del evento $A$.

Ejemplo del Teorema de Bayes aplicado a un problema real:

**Ejemplo.** Hay 100 monedas de las cuales una es **anormal**, con ambos lados cara (c).
Las restantes 99 monedas son normales con cara (c) en un lado y sello (s) en el otro.
Una moneda es tomada al azar y luego lanzada cinco veces seguidas.

En todas las cinco veces, la moneda cae mostrando solo el lado cara.
¿Cuál es la probabilidad de que esta moneda sea la **anormal**?

a)
*   Definamos $A$ el evento en el que la moneda es anormal.
*  Entonces $A^c$ es el evento en el que la moneda es normal.
*   Definamos también a $B$ como el evento en el cual la moneda cae cinco veces mostrando el lado cara.

b)

La respuesta está dada por $P(A|B)$:

<br>

\begin{align}
P(A|B)=\frac{P(B|A)P(A)}{P(B|A)P(A)+P(B|A^c)P(A^c)}=\frac{1 \times \frac{1}{100}}{1 \times \frac{1}{100}+(\frac{1}{2})^5 \times \frac{99}{100}} \approx \textbf{0.244}
\end{align}

<br>

Note que la respuesta no es lo que esperaríamos "intuitivamente".

El Teorema de Bayes "mezcla" el conocimiento anterior de que una de las 100 monedas es anormal con el dato de que solo una cara fue mostrada de los cinco lanzamientos.

Un tipo particular de cancer afecta el 0.1% de la población.

<br>

$\to$ Este método puede diagnosticar como positivo (+) en 99% de los casos verdaderos ($D$).

$\to$ Este método puede diagnosticar como negativo (-) en 95% de los casos falsos ($D^c$).

Este método de diagnóstico fue administrado a un paciente dando un resultado positivo (+). ¿Cuál es la probabilidad de que este paciente **realmente** tenga cancer?

<br>

a) Tenemos

$P(D)=0.001$

$P(D^c)=0.999$

$P(+|D^c)=1-P(-|D^c)=0.95$   $⇒$ **Sensibilidad 0.99**


Podemos también deducir que $P(+|D^c)=1-P(-|D^c)=0.05$   $⇒$ **especificidad 0.05**


<br>

b)
Aplicando el Teorema de Bayes:
La respuesta al problema está dada por $P(D|+)$:

$P(D|+)=\frac{P(+|D)P(D)}{P(+|D)P(D)+P(+|D^c)P(D^c)}=\frac{0.99 \times 0.001}{0.99 \times 0.001 + 0.05 \times 0.999} ≈ \textbf{0.02}$

Aún cuando la sensibilidad y especificidad eran "aparentemente" altas, a probabilidad $P(D|+) es sorpresivamente baja!.

La información anterior indica que en general la probabilidad de esta enfermedad es algo baja en 0.1%.

Con el fin de "superar" esta baja tasa de ocurrencia necesitamos un método de diagnóstico con una sensibilidad y especificidad más altas que las dadas aquí.

$\to$ Denotaremos en mayúscula la variable aleatoria y en minúscula un caso particular de el.

**Ejemplo.** Dada una variable aleatoria $X$ la probabilidad de que $X$ tome un valor $x$ está expresada por $P(X=x)$ o $P(x).

### **Población y Muestra**

**Población**

$\to$ El conjunto de datos completo es sujeto de análisis.

$\to$ Puede ser "real" o "idealizado".

$\to$ Las propiedades de la población son llamadas **parámetros**.

**Ejemplo.** La media, desviación estándar, varianza, etc. de una **población**.

<br>

**Muestra**

$\to$ Es un subconjunto de la población.

$\to$ Las propiedades de la muestra son llamadas **estadísticas**.

**Ejemplo.** La media, desviación estándar, varianza, etc. de una **muestra**.

<br>

Una pregunta común es "¿Qué tan grande puede ser una población?"

Recordemos la diferencia entre "población" vs "muestra" es principalmente conceptual.

Aún cuando un conjunto de datos es relativamente "pequeño", si representa a la totalidad entonces se puede calificar como "población".

Podemos calcular la media, la desviación estándar, la varianza, etc. usando la función de distribución de probabilidad.

$\to$ Estas cantidades son propiedades de un grupo "idealizado".

$\to$ Pueden ser interpretados como **parámetros** de una población **ideal**.

En otras palabras, las propiedades estadísticas de una población se llaman parámetros.

### **Media Poblacional o Media de la Población**

**Propiedades de la media poblacional**

<br>

$\to$ También es conocido como el **valor esperado** de una variable aleatoria: $E[X]$.

<br>

1) Para el caso discreto con $P(x)= \text{función de distribución de probabilidad}$

\begin{align}
\mu =E[X]=\sum_{\text{all } x}x P(x)
\end{align}


2) Para el caso contínuo con $f(x)= \text{función de densidad de probabilidad}$:

\begin{align}
\mu =E[X]=\int x f(x) dx
\end{align}

<br>

La media poblacional y el valor esperado de la variable aleatoria son iguales.

<br>

Para el caso discreto, es un promedio ponderado usando la probabilidad como "peso".

Para el caso contínuo, tenemos que integrar en vez de sumar.

<br>

**Propiedades de la media poblacional**

1) $E[c]=c$

2) $E[c X] = cE[X]$

3) $E[X + c]=E[X]+c$

4) $E[X + Y] = E[X]+E[Y]$

Nota: $c$ representa a una constante.

Estas relaciones son muy útiles para simplificar los cálculos.

### **Varianza de la Poblacion**

La varianza de la población se denota con $\sigma^2$ o $Var(X)$.

1) Para el caso discreto con $P(x)= \text{función de distribución de probabilidad}$:

\begin{align}
\sigma^2=Var(X)=\sum_{\text{all }x}(x-\mu)^2 P(x)
\end{align}

<br>

2) Para los casos contínuos con $f(x)=\text{función de densidad de probabilidad}$

\begin{align}
\sigma^2=Var(X)=\int(x-\mu)^2 f(x)dx
\end{align}

<br>


Nota: En ambos casos, $\sigma^2=E[X^2]-(E[X])^2$.

<br>

Nota: La desviación estándar de la población: $\sigma=\sqrt{\sigma^2}$

<br>

**Propiedades de la varianza de la población**

1) $Var(c)=0$

2) $Var(X+c)=Var(X)$

3)$Var(c X)=c^2 Var(x)$

4) \begin{align}
Var(X+Y)=Var(X)+Var(Y)+Cov(X, Y) \\
= Var(X)+Var(Y) \Leftarrow \text{Sólo cuando X y Y son independientes entre sí!}
\end{align}

<br>

Estas relaciones son útiles para simplificar los cálculos.

Aquí, $c$ se refiere a una constante cualquiera.

Cuando $c$ es sumada a una variable aleatoria, no contribuye a la varianza.

Cuando $c$ es multiplicada con una variable aleatoria, contribuye en $c^2$ multiplicado con la varianza.

Como la varianza de la suma de dos variables aleatorias, usted debe distinguir si estas variables son independientes o no.

**Variable aleatoria Poisson y distribución de probabilidad**

<br>

*  Es nombrado en honor al científico francés Simeon D. Poisson.

*  Describe la frecuencia de eventos para un intervalo dado de tiempo o espacio.

**Ejemplo.** Número de emails recibidos por hora.

**Ejemplo.** Número de terremotos por año.

**Ejemplo.** Número de chips de chocolate en una galleta.


<br>

La variable aleatoria de Poisson es otro ejemplo de variable aleatoria discreta.

<br>

1) "$X$ es una variable aleatorioa Poisson". $\Leftrightarrow X ~ Pois( \lambda )$

<br>

2) $P(x)=\frac{\lambda^x e^{-\lambda}}{x!} \Leftarrow 0 \le x$

*  Media = $\lambda$
*  Varianza = $\lambda$
*  Desviación estándar=$\sqrt{\lambda}$

<br>

3) Incrementando $n$ mientras decrece $p$, la distribución Binomial converge a la de Poisson con $\lambda=np$

<br>

Entre la Binomial y la Poisson hay similitudes y diferencias.

<br>

a) Similitudes:
*  Ambos modelan la frecuencia o conteo.
*  Ambos entregan la probabilidad para una frecuencia dada $x$.

<br>

b) Diferencias:
*  Para la Binomial, hay un límite superior para la frecuencia: $0 \le x \le n$

*  Para la Poisson, no hay límite superior para la frecuencia: $0 \le x$

<br>

Cuando $n$ se vuelve grande mientras mantenemos $p$ prqueño de modo que el límite superior se vuelve "irrelevante", entonces la Binomial "converge" a la Poisson.

```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

**1.1 Binomial:**

```python
# Tamaño de la muestra y el parámetro p
n=10
p=0.5
```

```python
# Distribución de probabilidad en x = 5
st.binom.pmf(5, n, p)
```

```python
# Cuantil en alpha = 0.3
# Más sobre cuantiles en la unidad III
st.binom.ppf(0.009, n, p)
```

```python
# P(0 <= x <= 5)
st.binom.cdf(5,n,p)
```

```python
# P(3 <= x <=7)
st.binom.cdf(7,n,p)-st.binom.cdf(2,n,p)
```

```python
# Visualizing the probability distribution.
x=np.arange(0,11)
plt.scatter(x, st.binom.pmf(x,n,p),color='red')
plt.show()
```
