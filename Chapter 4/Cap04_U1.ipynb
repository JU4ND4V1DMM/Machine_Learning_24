### **Teorema de Bayes**

Recordemos la siguiente relacin:

\begin{align}
P(A \cap B) = P(A|B)P(B) \\
=P(B|A)P(A)
\end{align}

<br>

Usando la relacin anterior, podemos mostrar el conocido **Teorema de Bayes**.

\begin{align}
P(A|B)P(B)=P(B|A)P(A)
\end{align}

<br>

Este teorema parece simple, pero hay una gran cantidad de consequencias a partir de este teorema.

Incluso existe una rama de la estadstica llamada "Estadstica Bayesiana".

Podemos transformar el Teorema de Bayes en la siguiente forma:

\begin{align}
P(A|B)P(B)=P(B|A)P(A)
\end{align}
\begin{align}
\bigg\downarrow
\end{align}

\begin{align}
P(A|B)=\frac{P(B|A)P(A)}{P(B)} \\
\frac{P(B|A)P(A)}{P(B|A)P(A)+P(B|A^c)P(A^c)}
\end{align}

Hemos transformado el Teorema de Bayes en un formato ms "til".

Preste especial atencin al denominador.

Por qu no calcular solo el $P(B)$ en el denominador sin adentrarnos en todo este problema?

Porque no se puede. Ya veremos por qu.

Podemos transformar el teorema de Bayes en la siguiente forma:

\begin{align}
P(A|B)=\frac{P(B|A)P(A)}{P(B|A)P(A)+P(B|A^c)P(A^c)}
\end{align}

### **Poblacin y Muestra**

### **Media Poblacional o Media de la Poblacin**

**Propiedades de la media poblacional**

<br>

$\to$ Tambin es conocido como el **valor esperado** de una variable aleatoria: $E[X]$.

<br>

1) Para el caso discreto con $P(x)= \text{funcin de distribucin de probabilidad}$

\begin{align}
\mu =E[X]=\sum_{\text{all } x}x P(x)
\end{align}


2) Para el caso contnuo con $f(x)= \text{funcin de densidad de probabilidad}$:

\begin{align}
\mu =E[X]=\int x f(x) dx
\end{align}

<br>

La media poblacional y el valor esperado de la variable aleatoria son iguales.

<br>

Para el caso discreto, es un promedio ponderado usando la probabilidad como "peso".

Para el caso contnuo, tenemos que integrar en vez de sumar.

<br>

**Propiedades de la media poblacional**

1) $E[c]=c$

2) $E[c X] = cE[X]$

3) $E[X + c]=E[X]+c$

4) $E[X + Y] = E[X]+E[Y]$

Nota: $c$ representa a una constante.

Estas relaciones son muy tiles para simplificar los clculos.

### **Varianza de la Poblacion**

La varianza de la poblacin se denota con $\sigma^2$ o $Var(X)$.

1) Para el caso discreto con $P(x)= \text{funcin de distribucin de probabilidad}$:

\begin{align}
\sigma^2=Var(X)=\sum_{\text{all }x}(x-\mu)^2 P(x)
\end{align}

<br>

2) Para los casos contnuos con $f(x)=\text{funcin de densidad de probabilidad}$

\begin{align}
\sigma^2=Var(X)=\int(x-\mu)^2 f(x)dx
\end{align}

<br>


Nota: En ambos casos, $\sigma^2=E[X^2]-(E[X])^2$.

<br>

Nota: La desviacin estndar de la poblacin: $\sigma=\sqrt{\sigma^2}$

<br>

**Propiedades de la varianza de la poblacin**

1) $Var(c)=0$

2) $Var(X+c)=Var(X)$

3)$Var(c X)=c^2 Var(x)$

4) \begin{align}
Var(X+Y)=Var(X)+Var(Y)+Cov(X, Y) \\
= Var(X)+Var(Y) \Leftarrow \text{Slo cuando X y Y son independientes entre s!}
\end{align}

<br>

Estas relaciones son tiles para simplificar los clculos.

Aqu, $c$ se refiere a una constante cualquiera.

Cuando $c$ es sumada a una variable aleatoria, no contribuye a la varianza.

Cuando $c$ es multiplicada con una variable aleatoria, contribuye en $c^2$ multiplicado con la varianza.

Como la varianza de la suma de dos variables aleatorias, usted debe distinguir si estas variables son independientes o no.

La variable aleatoria de Poisson es otro ejemplo de variable aleatoria discreta.

<br>

1) "$X$ es una variable aleatorioa Poisson". $\Leftrightarrow X ~ Pois( \lambda )$

<br>

2) $P(x)=\frac{\lambda^x e^{-\lambda}}{x!} \Leftarrow 0 \le x$

*  Media = $\lambda$
*  Varianza = $\lambda$
*  Desviacin estndar=$\sqrt{\lambda}$

<br>

3) Incrementando $n$ mientras decrece $p$, la distribucin Binomial converge a la de Poisson con $\lambda=np$

<br>

Entre la Binomial y la Poisson hay similitudes y diferencias.

<br>

a) Similitudes:
*  Ambos modelan la frecuencia o conteo.
*  Ambos entregan la probabilidad para una frecuencia dada $x$.

<br>

b) Diferencias:
*  Para la Binomial, hay un lmite superior para la frecuencia: $0 \le x \le n$

*  Para la Poisson, no hay lmite superior para la frecuencia: $0 \le x$

<br>

Cuando $n$ se vuelve grande mientras mantenemos $p$ prqueo de modo que el lmite superior se vuelve "irrelevante", entonces la Binomial "converge" a la Poisson.
