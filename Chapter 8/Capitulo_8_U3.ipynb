$\color{blue}{\to}$ Las capas se agregan a un objeto de la clase *`Sequential`*.

$\color{blue}{\to}$ Un objeto *`Model`* toma como argumentos la primera y la última capa.

$\color{blue}{\to}$ `Más flexible`.

<br>

$\color{blue}{\text{|}}$ **Lectura de los datos y exploración**

$\color{blue}{\to}$ Leyendo los datos

```python
## /content/drive/MyDrive/Samsung Innovation Campus 2023/CAP_08/
df = pd.read_csv('Datos/data_boston.csv', header='infer', encoding='latin1')
X = df.drop(columns=['PRICE'])
y = df['PRICE']
```

$\color{blue}{\to}$ Vista de los datos

```python
df.head(5)
```

```python
X.head(5)
```

$\color{blue}{\to}$ Escala los datos X.


```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

$\color{blue}{\to}$  Divide los datos en entrenamiento y evaluación

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
n_vars = X_train.shape[1]
```

$\color{blue}{\text{|}}$ **Definición de un modelo de API secuencial**


$\color{blue}{\to}$ Agrega capas a un objeto secuencial


```python
# agregando una capa de salida para regresión lineal
my_model1 = Sequential()
my_model1.add(Dense(input_dim = n_vars, units=1, activation="linear"))
```

```python
#resumen del modelo
my_model1.summary()
```

$\color{blue}{\text{|}}$ **Definición de los hiperparámetros y el optimizador**


```python
#hiperparámetros
n_epochs = 2000
batch_size = 10
learn_rate = 0.002
```

```python
#Define el optimizador y luego compila
my_optimizer=Adam(learning_rate=learn_rate)
my_model1.compile(loss = "mse", optimizer = my_optimizer, metrics=["mse", "mape"])  ## "R2Score" no disponible en tf 2.10.0
```

$\color{blue}{\text{|}}$ **Entrenamiento del modelo y visualización del historial**

$\color{blue}{\to}$ Entrena el modelo.

```python
print("tensorflow version: ", tf.__version__)
physical_devices = tf.config.list_physical_devices('GPU')

print("GPU:", tf.config.list_physical_devices('GPU'))
```

```python
%%time
#verbose=0 significa sin visualización del progreso de entrenamiento
#verbose=1 para ver las épocas de entrenamiento
my_summary = my_model1.fit(X_train, y_train, epochs=n_epochs, batch_size = batch_size,
                           validation_split = 0.2, verbose = 1) ## 8min 54s local GPU(NVIDIA GeForce GTX 3050)
```

```python
my_summary.history.keys()  #mostrando los keys
```

$\color{blue}{\to}$ Visualiza el historial de entrenamiento.


```python
n_skip = 100
plt.plot(my_summary.history['mse'][n_skip:], c="b")
plt.plot(my_summary.history['val_mse'][n_skip:], c="g")
plt.title('Historial de Entrenamiento')
plt.ylabel('MSE')
plt.xlabel('Epoca')
plt.legend(['Entrenamiento', 'Validación'], loc='upper right')
plt.show()
```

$\color{blue}{\text{|}}$ **Evaluación**

$\color{blue}{\to}$ Predicción y evaluación usando una fórmula

```python
y_pred = my_model1.predict(X_test)[:, 0]
RMSE = np.sqrt(np.mean((y_test-y_pred)**2))
np.round(RMSE, 3)
```

$\color{blue}{\to}$ Usa el método *evaluate()*

```python
MSE = my_model1.evaluate(X_test, y_test, verbose=0)[1]
RMSE = np.sqrt(MSE)
print("RMSE de la evaluación : {}".format(np.round(RMSE,3)))
```

#### **Diagrama de una red neuronal convolucional (CNN)**

Las capas de convolución y agrupación se repiten varias veces.
Hay una capa completamente conectada justo antes de la salida.

<br>

$\color{blue}{\to}$ Aquí, un diagrama muestra la estructura de una CNN

SIC2023_C8_U3_189.png

$\color{blue}{\to}$ Observamos que en la **"fase de extracción de atributos"** la convolución y la agrupación se repiten varias veces.

$\color{blue}{\to}$ En la **"fase de clasificación"**, suele haber una capa de descarte o *abandono* y una capa "aplanada" totalmente conectada.

$\color{blue}{\to}$ Suponemos que la variable de respuesta ha sido codificada en one-hot-encoding.

$\color{blue}{\to}$ El "argmax" de la salida da la predicción real en términos de las etiquetas (0~9).


<br>

$\color{blue}{\text{|}}$ Leyendo los datos

```python
# librerías necesarias

import numpy as np
import matplotlib.pyplot as plt
import warnings

from sklearn.model_selection import train_test_split

from tensorflow.keras.datasets.mnist import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D

from tensorflow.keras.optimizers import Adam, RMSprop, SGD
warnings.filterwarnings('ignore')  #desactivando advertencias del sistema
%matplotlib inline
```

$\color{blue}{\to}$ Cargando los datos

```python
(X_train, y_train), (X_test, y_test) = load_data()
n_train_size = X_train.shape[0]
```

$\color{blue}{\to}$ Viendo las formas

```python
print("-"*50)
print("Forma de los datos X de entrenamiento: {}".format(X_train.shape))
print("Forma de los datos y de entrenamiento: {}".format(y_train.shape))
print("-"*50)
print("Forma de los datos X de evaluación: {}".format(X_test.shape))
print("Forma de los datos y de evaluación: {}".format(y_test.shape))
print("-"*50)
```

$\color{blue}{\text{|}}$ Visualizando los datos

```python
image = 0 #X_train tiene varias imagenes, con este numero seleccionamos cual imagen mostrar

plt.imshow(X_train[image,:,:],cmap="Greys")
plt.show()
```

$\color{blue}{\to}$ Escalando

```python
X_train = X_train/255
X_test = X_test/255
```

$\color{blue}{\to}$ Ajustando la forma

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
image = 0 #X_train tiene varias imagenes, con este numero seleccionamos cual imagen mostrar

plt.imshow(X_train[image,:,:],cmap="Greys")
plt.show()
```

```python
X_train.shape
```

$\color{blue}{\to}$ One-Hot-Encoding

```python
y = np.concatenate([y_train, y_test], axis=0)
y = tf.keras.utils.to_categorical(y, 10)
y_train = y[:n_train_size,:]
y_test = y[n_train_size:,:]
```

```python
y_test
```

$\color{blue}{\text{|}}$ **Definiendo de un modelo Red Neuronal Convolucional (CNN)**


```python
drop_prob = 0.5
my_model = Sequential()

# Primera convolución + agrupación
my_model.add(Conv2D(input_shape=(28,28,1), filters=32, kernel_size=(5, 5), padding='same',
             activation="relu"))
my_model.add(MaxPooling2D(pool_size=2))

# Segunda convolución + agrupación
my_model.add(Conv2D(filters=64, kernel_size=(5, 5), padding='same', activation="relu"))
my_model.add(MaxPooling2D(pool_size=2))

# Capa aplanada y totalmente conectada
my_model.add(Flatten())
my_model.add(Dense(units = 1024, activation="relu"))

# Aplicando Abandono
my_model.add(Dropout(rate=drop_prob))

# Capa de salida
my_model.add(Dense(units = 10, activation="softmax"))
```

$\color{blue}{\to}$ Ver el resumen

```python
my_model.summary()
```

$\color{blue}{\text{|}}$ Definición de los hiperparámetros y el optimizador

```python
n_epochs = 10
batch_size = 200
learn_rate = 0.001
```

```python
my_optimizer=Adam(learning_rate=learn_rate)
my_model.compile(loss = "categorical_crossentropy", optimizer = my_optimizer, metrics=["accuracy"])
```

$\color{blue}{\text{|}}$ Entrenamiento del modelo y visualización del historial


```python
# a continuación entrenamos el modelo, notemos que verbose lo colocamos en 1 y podremos ver el progreso
# del entrenamoiento

my_summary = my_model.fit(X_train, y_train, epochs=n_epochs, batch_size = batch_size,
                          validation_split = 0.2, verbose = 1)
```

```python
# graficamos el histórico del entrenamiento

n_skip = 0
plt.plot(my_summary.history['accuracy'][n_skip:], c="b")
plt.plot(my_summary.history['val_accuracy'][n_skip:], c="g")
plt.title('Historial de Entrenamiento')
plt.ylabel('Accuracy')
plt.xlabel('Epoca')
plt.legend(['Entrenamiento', 'Validación'], loc='upper right')
plt.show()
```

$\color{blue}{\text{|}}$ Evaluación

```python
ACC = my_model.evaluate(X_test, y_test, verbose=0)[1]
print("Accuracy de la evaluación : {}".format(np.round(ACC,3)))
```

```python
# librerías necesarias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

import tensorflow as tf
from sklearn.model_selection import train_test_split

from tensorflow.keras.datasets.cifar10 import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D

from tensorflow.keras.optimizers import Adam, RMSprop, SGD
warnings.filterwarnings('ignore')  #desactivando advertencias del sistema
%matplotlib inline
```

$\color{blue}{\to}$ Cargando los datos

```python
(X_train, y_train), (X_test, y_test) = load_data()
n_train_size = X_train.shape[0]
```

$\color{blue}{\to}$ Viendo las formas

```python
print("-"*50)
print("Forma de los datos X de entrenamiento: {}".format(X_train.shape))
print("Forma de los datos y de entrenamiento: {}".format(y_train.shape))
print("-"*50)
print("Forma de los datos X de evaluación: {}".format(X_test.shape))
print("Forma de los datos y de evaluación: {}".format(y_test.shape))
print("-"*50)
```

$\color{blue}{\text{|}}$ Visualizando los datos

```python
image = 123  #el dataset tiene varias imagenes, con este valor escogemos cual mostrar
plt.imshow(X_train[image,:,:])
plt.show()
```

$\color{blue}{\text{|}}$ Preparación de los datos

$\color{blue}{\to}$ Escalando

```python
X_train = X_train/255
X_test = X_test/255
```

$\color{blue}{\to}$ Ajustando la forma de los datos

```python
X_train = X_train.reshape(-1,32,32,3)
X_test = X_test.reshape(-1,32,32,3)
```

$\color{blue}{\to}$ One-hot-encoding

```python
y = np.concatenate([y_train,y_test],axis=0)
y = tf.keras.utils.to_categorical(y,10)
y_train = y[:n_train_size,:]
y_test = y[n_train_size:,:]
```

$\color{blue}{\text{|}}$ Definiendo un modelo CNN

```python
drop_prob = 0.7
my_model = Sequential()

# Primera convolución + agrupación
my_model.add(Conv2D(input_shape=(32,32,3), filters=32, kernel_size=(5, 5), padding='same',
             activation="relu"))      #3 canales de color

my_model.add(MaxPooling2D(pool_size=2))

# solo la primera capa es distinta en este caso de 3 canales de color
# las demas capas se mantienen con la misma configuración

# Segunda convolución + agrupación
my_model.add(Conv2D(filters=64, kernel_size=(5, 5), padding='same', activation="relu"))
my_model.add(MaxPooling2D(pool_size=2))

# Capa aplanada y totalmente conectada
my_model.add(Flatten())
my_model.add(Dense(units = 1024, activation="relu"))

# Aplicando Abandono
my_model.add(Dropout(rate=drop_prob))

# Capa de salida
my_model.add(Dense(units = 10, activation="softmax"))
```

```python
my_model.summary()
```

$\color{blue}{\text{|}}$ Definiendo los hiperparámetros y el optimizador


```python
n_epochs = 20
batch_size = 20
learn_rate = 0.0001
```

```python
my_optimizer=Adam(learning_rate=learn_rate)
my_model.compile(loss="categorical_crossentropy", optimizer=my_optimizer, metrics=["accuracy"])
```

$\color{blue}{\text{|}}$ Entrenamiento del modelo y visualización del historial

```python
my_summary = my_model.fit(X_train, y_train, epochs=n_epochs,
                          batch_size=batch_size, validation_split=0.3, verbose=1)
```

```python
# graficamos el histórico del entrenamiento

n_skip = 0
plt.plot(my_summary.history['accuracy'][n_skip:], c="b")
plt.plot(my_summary.history['val_accuracy'][n_skip:], c="g")
plt.title('Historial de Entrenamiento')
plt.ylabel('Accuracy')
plt.xlabel('Epoca')
plt.legend(['Entrenamiento', 'Validación'], loc='upper right')
plt.show()
```

$\color{blue}{\text{|}}$ Evaluación

```python
ACC = my_model.evaluate(X_test, y_test, verbose=0)[1]
print("Accuracy de la evaluación : {}".format(np.round(ACC, 3)))
```

$\color{blue}{\to}$ Leyendo los datos

```python
df = pd.read_csv('/content/drive/My Drive/Ejercicio_Febrero_2024/data_time_series.csv', header='infer', encoding='latin1')
n_times_steps = df.shape[0]
print(df.shape)
```

$\color{blue}{\to}$ Mostrando el DataFrame

```python
df.head()
```

$\color{blue}{\to}$ Vizualiza la serie de tiempo

```python
plt.plot(df.Value, c="b", linewidth=2, linestyle="-")
plt.show()
```

$\color{blue}{\to}$ Modelo de suavizado exponencial multiplicativo

```python
#modelo de suavizado de datos
model = ExponentialSmoothing(df['Value'][0:40], seasonal='mul', seasonal_periods=6).fit()

#se ejecuta el modelo, suavizando los datos
pred = model.predict(start=0, end=50)

#se grafica el resultado comparando en azul sin suavizar y en rojo ya suavizado
plt.plot(df, c='b', linewidth=2, linestyle='-', label='Datos')
plt.plot(pred, c='r', linewidth=2, linestyle='--', label='Ajustado')
```

$\color{blue}{\to}$ Preprocesamiento

$\color{blue}{\to}$ Escalar datos (presta atención a la escala vertical del gráfico)


```python
scaler = MinMaxScaler()
ts_scaled = scaler.fit_transform(df)
plt.plot(ts_scaled, c='g', linewidth=2, linestyle="-")
plt.show()
```

$\color{blue}{\to}$ Ajustando la forma de los datos (batch_size, duración de series temporales, n_input)


```python
ts_scaled_2 = ts_scaled.reshape(1, -1, 1)

ts_scaled_2.shape
```

$\color{blue}{\text{|}}$ Definiciones necesarias

$\color{blue}{\to}$ Hiperparámetros

```python
n_epochs = 1001
batch_size = 1  # solo hay una serie de tiempo, entonces no hay otra opción que 1
learn_rate = 0.0002
```

```python
n_input = 1      # entrada escalar
n_neurons = 100  # cantidad de neuronas por capa
n_output = 1     # salida escalar
```

<br>

$\color{blue}{\text{|}}$ Red RNN o LSTM

```python
my_model = Sequential()
my_model.add(SimpleRNN(units=n_neurons, return_sequences=True, input_shape=(None, n_input)))

my_model.add(TimeDistributed(Dense(units=n_output, activation="linear")))
```

$\color{blue}{\to}$ Ver el resumen

```python
my_model.summary()
```

$\color{blue}{\to}$ Define el optimizador y compila

```python
my_optimizer=Adam(learning_rate=learn_rate)
my_model.compile(loss="mse", optimizer=my_optimizer, metrics=["mse"])
```

$\color{blue}{\to}$ Entrena el modelo

```python
my_summary = my_model.fit(ts_scaled_2[:, :-1, :], ts_scaled_2[:, 1:, :], epochs=n_epochs,
                          batch_size=batch_size, verbose=0)

#ten en cuenta que y es X corrido en +1
```

```python
my_summary.history.keys()
```

$\color{blue}{\text{|}}$ Graficando el historial de entrenamiento

```python
plt.plot(my_summary.history['mse'], c="b")
plt.title('Historial de Entrenamiento')
plt.ylabel('MSE')
plt.xlabel('Epoca')
plt.legend(['Entrenamiento'], loc='upper right')
plt.show()
```

$\color{blue}{\to}$ Predice el futuro

```python
n_ts_seed = 5   # tamaño de la semilla
n_predict_time_steps = 55  # tamaño de la predicción
```

$\color{blue}{\to}$ Bucle de predicción

```python
ts_seed = ts_scaled[0:n_ts_seed]
for i in range(n_predict_time_steps):
  X = ts_seed.reshape(1, -1, 1)
  y_pred = my_model.predict(X, verbose=0)
  y_last = y_pred[0, -1, 0]
  ts_seed = np.concatenate((ts_seed, np.array([y_last]).reshape(1, 1)), axis=0)
```

$\color{blue}{\to}$ Haz la transformación inversa y muestra las series temporales

```python
ts = scaler.inverse_transform(ts_seed)
plt.plot(df.Value, c='b', linewidth=2, linestyle="-", label="Datos")
plt.plot(ts, c='r', linewidth=2, linestyle="--", label="Ajuste")
plt.legend()
plt.tight_layout()
plt.savefig("out.png")
plt.show()
```
